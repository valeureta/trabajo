```python
def qident(s: str) -> str:
    return f"`{s.replace('`','``')}`"

def qtable(full: str) -> str:
    return ".".join(qident(p) for p in full.split("."))

def preparar_bd():
    conn = mysql.connector.connect(
        host=MYSQL_HOST, port=MYSQL_PORT,
        user=MYSQL_USER, password=MYSQL_PASSWORD,
        database=MYSQL_DB,
        allow_local_infile=True,
    )
    try:
        cur = conn.cursor()
        try:
            cur.execute(f"ALTER TABLE {TABLE_FINAL} ADD COLUMN Fecha DATE;")
        except mysql.connector.Error:
            pass
        try:
            cur.execute(f"ALTER TABLE {TABLE_FINAL} ADD COLUMN `Hora 25` DECIMAL(18,6) NULL;")
        except mysql.connector.Error:
            pass

        cur.execute(f"CREATE TABLE IF NOT EXISTS {TABLE_STG} LIKE {TABLE_FINAL};")
        try:
            cur.execute(f"ALTER TABLE {TABLE_FINAL} ADD UNIQUE KEY uq_llave_fecha (Llave, Fecha);")
        except mysql.connector.Error:
            pass

        conn.commit()
    finally:
        conn.close()

def truncate_staging(cur):
    cur.execute(f"TRUNCATE TABLE {TABLE_STG};")

def contar_columnas(csv_path: Path) -> int:
    with open(csv_path, "r", encoding="utf-8", newline="") as f:
        r = csv.reader(f, delimiter=";")
        header = next(r)
    return len(header)

def load_to_staging(cur, csv_path: Path):
    n = contar_columnas(csv_path)

    if n == 33:
        cols = """
        (
          @ano, @mes,
          `Llave`, `Central`, `Coordinado`, `Grupo reporte`, `Tipo`, `Subtipo`,
          @fecha_str,
          @h1,@h2,@h3,@h4,@h5,@h6,@h7,@h8,@h9,@h10,@h11,@h12,@h13,@h14,@h15,@h16,@h17,@h18,@h19,@h20,@h21,@h22,@h23,@h24
        )
        """
        set_h25 = "`Hora 25` = NULL"
    elif n == 34:
        cols = """
        (
          @ano, @mes,
          `Llave`, `Central`, `Coordinado`, `Grupo reporte`, `Tipo`, `Subtipo`,
          @fecha_str,
          @h1,@h2,@h3,@h4,@h5,@h6,@h7,@h8,@h9,@h10,@h11,@h12,@h13,@h14,@h15,@h16,@h17,@h18,@h19,@h20,@h21,@h22,@h23,@h24,@h25
        )
        """
        set_h25 = "`Hora 25` = NULLIF(REPLACE(@h25, ',', '.'),'')"
    else:
        raise ValueError(f"{csv_path.name}: columnas inesperadas ({n}).")

    sql = f"""
    LOAD DATA LOCAL INFILE %s
    INTO TABLE {TABLE_STG}
    CHARACTER SET utf8mb4
    FIELDS TERMINATED BY ';'
    LINES TERMINATED BY '\\n'
    IGNORE 1 LINES
    {cols}
    SET
      Fecha = STR_TO_DATE(@fecha_str, '{CSV_FECHA_FMT}'),
      `Hora 1`  = NULLIF(REPLACE(@h1,  ',', '.'),''),
      `Hora 2`  = NULLIF(REPLACE(@h2,  ',', '.'),''),
      `Hora 3`  = NULLIF(REPLACE(@h3,  ',', '.'),''),
      `Hora 4`  = NULLIF(REPLACE(@h4,  ',', '.'),''),
      `Hora 5`  = NULLIF(REPLACE(@h5,  ',', '.'),''),
      `Hora 6`  = NULLIF(REPLACE(@h6,  ',', '.'),''),
      `Hora 7`  = NULLIF(REPLACE(@h7,  ',', '.'),''),
      `Hora 8`  = NULLIF(REPLACE(@h8,  ',', '.'),''),
      `Hora 9`  = NULLIF(REPLACE(@h9,  ',', '.'),''),
      `Hora 10` = NULLIF(REPLACE(@h10, ',', '.'),''),
      `Hora 11` = NULLIF(REPLACE(@h11, ',', '.'),''),
      `Hora 12` = NULLIF(REPLACE(@h12, ',', '.'),''),
      `Hora 13` = NULLIF(REPLACE(@h13, ',', '.'),''),
      `Hora 14` = NULLIF(REPLACE(@h14, ',', '.'),''),
      `Hora 15` = NULLIF(REPLACE(@h15, ',', '.'),''),
      `Hora 16` = NULLIF(REPLACE(@h16, ',', '.'),''),
      `Hora 17` = NULLIF(REPLACE(@h17, ',', '.'),''),
      `Hora 18` = NULLIF(REPLACE(@h18, ',', '.'),''),
      `Hora 19` = NULLIF(REPLACE(@h19, ',', '.'),''),
      `Hora 20` = NULLIF(REPLACE(@h20, ',', '.'),''),
      `Hora 21` = NULLIF(REPLACE(@h21, ',', '.'),''),
      `Hora 22` = NULLIF(REPLACE(@h22, ',', '.'),''),
      `Hora 23` = NULLIF(REPLACE(@h23, ',', '.'),''),
      `Hora 24` = NULLIF(REPLACE(@h24, ',', '.'),''),
      {set_h25};
    """
    cur.execute(sql, (str(csv_path.resolve()),))

def insert_ignore_final(cur):
    sql = f"""
    INSERT IGNORE INTO {TABLE_FINAL}
    (`Llave`, `Central`, `Coordinado`, `Grupo reporte`, `Tipo`, `Subtipo`, `Fecha`,
     `Hora 1`,`Hora 2`,`Hora 3`,`Hora 4`,`Hora 5`,`Hora 6`,`Hora 7`,`Hora 8`,`Hora 9`,`Hora 10`,
     `Hora 11`,`Hora 12`,`Hora 13`,`Hora 14`,`Hora 15`,`Hora 16`,`Hora 17`,`Hora 18`,`Hora 19`,`Hora 20`,
     `Hora 21`,`Hora 22`,`Hora 23`,`Hora 24`,`Hora 25`)
    SELECT
     s.`Llave`, s.`Central`, s.`Coordinado`, s.`Grupo reporte`, s.`Tipo`, s.`Subtipo`, s.`Fecha`,
     s.`Hora 1`,s.`Hora 2`,s.`Hora 3`,s.`Hora 4`,s.`Hora 5`,s.`Hora 6`,s.`Hora 7`,s.`Hora 8`,s.`Hora 9`,s.`Hora 10`,
     s.`Hora 11`,s.`Hora 12`,s.`Hora 13`,s.`Hora 14`,s.`Hora 15`,s.`Hora 16`,s.`Hora 17`,s.`Hora 18`,s.`Hora 19`,s.`Hora 20`,
     s.`Hora 21`,s.`Hora 22`,s.`Hora 23`,s.`Hora 24`,s.`Hora 25`
    FROM {TABLE_STG} s
    WHERE s.`Fecha` IS NOT NULL;
    """
    cur.execute(sql)

def subir_archivos(archivos: list[Path]):
    preparar_bd()

    conn = mysql.connector.connect(
        host=MYSQL_HOST, port=MYSQL_PORT,
        user=MYSQL_USER, password=MYSQL_PASSWORD,
        database=MYSQL_DB,
        allow_local_infile=True,
    )
    try:
        conn.autocommit = False
        cur = conn.cursor()

        cur.execute("SELECT DATABASE();")
        print("DB:", cur.fetchone()[0])

        cur.execute(f"SELECT COUNT(*) FROM {TABLE_FINAL};")
        print("FINAL antes:", cur.fetchone()[0])

        for p in archivos:
            p = Path(p)
            if not p.exists():
                raise FileNotFoundError(p)

            print("Subiendo:", p.name, "| columnas:", contar_columnas(p))
            truncate_staging(cur)
            load_to_staging(cur, p)

            cur.execute(f"SELECT COUNT(*), SUM(Fecha IS NULL), SUM(Fecha IS NOT NULL) FROM {TABLE_STG};")
            print("STG total / Fecha NULL / Fecha OK:", cur.fetchone())

            insert_ignore_final(cur)
            conn.commit()

        cur.execute(f"SELECT COUNT(*) FROM {TABLE_FINAL};")
        print("FINAL después:", cur.fetchone()[0])
        print("OK")

    except Exception:
        conn.rollback()
        raise
    finally:
        conn.close()

def column_exists(cur, schema: str, table: str, column: str) -> bool:
    cur.execute(
        """
        SELECT COUNT(*)
        FROM information_schema.COLUMNS
        WHERE TABLE_SCHEMA=%s AND TABLE_NAME=%s AND COLUMN_NAME=%s
        """,
        (schema, table, column),
    )
    return cur.fetchone()[0] > 0

def update_subtipo_gas():
    db, tbl = TABLE_OR.split(".", 1)

    conn = mysql.connector.connect(**MYSQL)
    try:
        cur = conn.cursor()

        # Validaciones rápidas de columnas
        if not column_exists(cur, db, tbl, COL_SUBTIPO):
            raise ValueError(f"No existe columna {COL_SUBTIPO} en {TABLE_OR}")
        if not column_exists(cur, db, tbl, COL_LLAVE):
            raise ValueError(f"No existe columna {COL_LLAVE} en {TABLE_OR}")

        # 1) GLP
        sql_glp = f"""
        UPDATE {TABLE_OR}
        SET {qident(COL_SUBTIPO)} = 'GLP'
        WHERE {qident(COL_SUBTIPO)} = 'Gas Natural'
          AND UPPER({qident(COL_LLAVE)}) LIKE '%%GLP%%';
        """
        cur.execute(sql_glp)
        glp_rows = cur.rowcount
        conn.commit()

        # 2) GNL
        sql_gnl = f"""
        UPDATE {TABLE_OR}
        SET {qident(COL_SUBTIPO)} = 'GNL'
        WHERE {qident(COL_SUBTIPO)} = 'Gas Natural'
          AND UPPER({qident(COL_LLAVE)}) LIKE '%%GNL%%';
        """
        cur.execute(sql_gnl)
        gnl_rows = cur.rowcount
        conn.commit()

        # 3) GNA (Gas Arg / Gas Argentino)
        sql_gna = f"""
        UPDATE {TABLE_OR}
        SET {qident(COL_SUBTIPO)} = 'GNA'
        WHERE {qident(COL_SUBTIPO)} = 'Gas Natural'
          AND (
                UPPER({qident(COL_LLAVE)}) LIKE '%%GAS ARGENTINO%%'
             OR UPPER({qident(COL_LLAVE)}) LIKE '%%GAS ARG%%'
          );
        """
        cur.execute(sql_gna)
        gna_rows = cur.rowcount
        conn.commit()

        # Resumen
        cur.execute(
            f"""
            SELECT {qident(COL_SUBTIPO)}, COUNT(*)
            FROM {TABLE_OR}
            WHERE {qident(COL_SUBTIPO)} IN ('Gas Natural','GNA','GNL','GLP')
            GROUP BY {qident(COL_SUBTIPO)}
            ORDER BY {qident(COL_SUBTIPO)};
            """
        )
        resumen = cur.fetchall()

        print("Actualizaciones:")
        print("  GLP:", glp_rows)
        print("  GNL:", gnl_rows)
        print("  GNA:", gna_rows)
        print("\nConteo final (Gas Natural / GNA / GNL / GLP):")
        for subt, cnt in resumen:
            print(f"  {subt}: {cnt}")

    finally:
        conn.close()
        
def build_unpivot_union(n_horas: int) -> str:
    # genera filas (Fecha, hora, Subtipo, valor) con UNION ALL
    parts = []
    for h in range(1, n_horas + 1):
        hora_out = h - 1  # Hora 1 -> hora 0
        parts.append(f"""
        SELECT
          Fecha,
          {hora_out} AS hora,
          Subtipo,
          COALESCE(`Hora {h}`, 0) AS valor
        FROM {TABLE_OR}
        WHERE Fecha BETWEEN %s AND %s
        """.strip())
    return "\nUNION ALL\n".join(parts)

def build_insert_sql(n_horas: int) -> str:
    unpivot = build_unpivot_union(n_horas)

    # columnas de la tabla subtipo (con backticks para las que tienen espacio)
    cols_insert = ["Fecha", "hora"] + [f"`{s}`" if " " in s else s for s in SUBTIPOS_OR]

    # expresiones SUM(CASE...) para cada subtipo
    sums = []
    for s in SUBTIPOS_OR:
        col_alias = f"`{s}`" if " " in s else s
        sums.append(f"SUM(CASE WHEN Subtipo = '{s}' THEN valor ELSE 0 END) AS {col_alias}")

    # updates ON DUPLICATE KEY
    updates = []
    for s in SUBTIPOS_OR:
        col = f"`{s}`" if " " in s else s
        updates.append(f"{col}=VALUES({col})")

    sql = f"""
    INSERT INTO {TABLE_SUB}
    ({", ".join(cols_insert)})
    SELECT
      Fecha,
      hora,
      {", ".join(sums)}
    FROM (
      {unpivot}
    ) u
    GROUP BY Fecha, hora
    ON DUPLICATE KEY UPDATE
      {", ".join(updates)};
    """
    return sql

def upsert_subtipo_all_hours(fecha_inicio, fecha_fin, n_horas: int):
    sql = build_insert_sql(n_horas)

    conn = mysql.connector.connect(
        host=MYSQL_HOST, port=MYSQL_PORT,
        user=MYSQL_USER, password=MYSQL_PASSWORD,
        database=MYSQL_DB,
    )
    try:
        cur = conn.cursor()
        params = []
        for _ in range(n_horas):
            params.extend([fecha_inicio, fecha_fin])

        cur.execute(sql, tuple(params))
        conn.commit()

        cur.execute(
            f"SELECT COUNT(*) FROM {TABLE_SUB} WHERE Fecha BETWEEN %s AND %s;",
            (fecha_inicio, fecha_fin)
        )
        print("Filas en subtipo (rango):", cur.fetchone()[0])
    finally:
        conn.close()

def q_literal(s: str) -> str:
    return s.replace("'", "''")

def build_grouping_map(centrales, agrupaciones_manuales):
    agrup = {}

    for c in centrales:
        if c in agrupaciones_manuales:
            agrup[c] = list(agrupaciones_manuales[c])
        else:
            agrup[c] = [c]

    return agrup

def validate_grouping_map(centrales, agrupaciones_manuales):
    errores = []

    for central_salida in agrupaciones_manuales:
        if central_salida not in centrales:
            errores.append(f"La central de salida '{central_salida}' no existe en CENTRALES.")

    todas_las_llaves = []
    for _, llaves in agrupaciones_manuales.items():
        todas_las_llaves.extend(llaves)

    repetidas = [k for k, n in Counter(todas_las_llaves).items() if n > 1]
    if repetidas:
        errores.append("Hay llaves repetidas en AGRUPACIONES_MANUALES: " + ", ".join(repetidas))

    if errores:
        raise ValueError("\n".join(errores))

def ensure_out_table(cur, table_out: str, centrales: list[str]):
    cols = ",\n  ".join([f"{qident(c)} DECIMAL(18,6) NULL" for c in centrales])
    cur.execute(f"""
    CREATE TABLE IF NOT EXISTS {table_out} (
      Fecha DATE NOT NULL,
      hora  TINYINT NOT NULL,
      {cols},
      PRIMARY KEY (Fecha, hora)
    ) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4 COLLATE=utf8mb4_0900_ai_ci;
    """)
    for c in centrales:
        try:
            cur.execute(f"ALTER TABLE {table_out} ADD COLUMN {qident(c)} DECIMAL(18,6) NULL;")
        except mysql.connector.Error:
            pass

def print_missing_grouped_keys(cur, fecha_inicio, fecha_fin, grouping_map):
    """
    Revisa si faltan llaves de las agrupaciones manuales en operacion_real.
    """
    llaves_esperadas = sorted({x for vals in grouping_map.values() for x in vals})
    if not llaves_esperadas:
        return

    placeholders = ", ".join(["%s"] * len(llaves_esperadas))
    sql = f"""
        SELECT DISTINCT o.`Llave`
        FROM {TABLE_OR} o
        WHERE o.Fecha BETWEEN %s AND %s
          AND o.`Llave` IN ({placeholders})
    """
    params = [fecha_inicio, fecha_fin, *llaves_esperadas]
    cur.execute(sql, params)
    found = {r[0] for r in cur.fetchall() if r and r[0] is not None}

    missing = [k for k in llaves_esperadas if k not in found]
    if missing:
        print(f"[WARN] Llaves esperadas no encontradas en {TABLE_OR} para el rango: {len(missing)}")
        for k in missing:
            print("  -", k)

def build_sum_case_for_output_col(output_col: str, source_keys: list[str], col_hora: str) -> str:
    condiciones = " OR ".join(
        f"o.`Llave` = '{q_literal(k)}'"
        for k in source_keys
    )
    return (
        f"SUM(CASE WHEN ({condiciones}) "
        f"THEN COALESCE(o.`{col_hora}`,0) ELSE 0 END) AS {qident(output_col)}"
    )

def upsert_centrales_all_hours(fecha_inicio, fecha_fin, n_horas: int):
    validate_grouping_map(CENTRALES, AGRUPACIONES_MANUALES)

    grouping_map = build_grouping_map(CENTRALES, AGRUPACIONES_MANUALES)
    llaves_absorbidas = sorted({x for vals in grouping_map.values() for x in vals})

    centrales_out = list(CENTRALES) + [OTROS_COL]

    conn = mysql.connector.connect(
        host=MYSQL_HOST, port=MYSQL_PORT,
        user=MYSQL_USER, password=MYSQL_PASSWORD,
        database=MYSQL_DB,
    )
    try:
        cur = conn.cursor()

        print_missing_grouped_keys(cur, fecha_inicio, fecha_fin, grouping_map)
        ensure_out_table(cur, TABLE_CEN, centrales_out)

        cols_insert = ", ".join(qident(c) for c in centrales_out)
        updates = ",\n  ".join(f"{qident(c)} = VALUES({qident(c)})" for c in centrales_out)

        in_list_otros = ", ".join(["%s"] * len(llaves_absorbidas))

        for h in range(1, n_horas + 1):
            hora_out = h - 1
            col_hora = f"Hora {h}"

            cols_select_main = ",\n  ".join(
                build_sum_case_for_output_col(c, grouping_map[c], col_hora)
                for c in CENTRALES
            )

            col_select_otros = (
                f"SUM(CASE "
                f"WHEN o.`Llave` IS NOT NULL AND o.`Llave` <> '' "
                f"  AND o.{qident(COL_TIPO)} = %s "
                f"  AND o.`Llave` NOT IN ({in_list_otros}) "
                f"THEN COALESCE(o.`{col_hora}`,0) ELSE 0 END) AS {qident(OTROS_COL)}"
            )

            cols_select = cols_select_main + ",\n  " + col_select_otros

            sql = f"""
            INSERT INTO {TABLE_CEN} (Fecha, hora, {cols_insert})
            SELECT
              o.Fecha,
              {hora_out} AS hora,
              {cols_select}
            FROM {TABLE_OR} o
            WHERE o.Fecha BETWEEN %s AND %s
            GROUP BY o.Fecha
            ON DUPLICATE KEY UPDATE
              {updates};
            """

            params = [TIPO_TERMICO, *llaves_absorbidas, fecha_inicio, fecha_fin]

            cur.execute(sql, params)
            conn.commit()
            print(f"Hora {h} OK")

        cur.execute(
            f"SELECT COUNT(*) FROM {TABLE_CEN} WHERE Fecha BETWEEN %s AND %s;",
            (fecha_inicio, fecha_fin)
        )
        print("Filas en datos_operacion_real.centrales (rango):", cur.fetchone()[0])

    finally:
        conn.close()
```

```python
# CELDA — Pivot por Subtipo (auto-detecta subtipos y columnas Hora) y upsert a planificacion.subtipos
import mysql.connector
from datetime import date

# ---- CONFIG ----
MYSQL_HOST = "127.0.0.1"
MYSQL_PORT = 3306
MYSQL_USER = "root"
MYSQL_PASSWORD = "rootvale123"

MYSQL_DB = "costos_variables"
TABLE_OR = "costos_variables.prg_energia_horaria"
TABLE_OUT = "planificacion.subtipos"

# ---------------

def q_ident(s: str) -> str:
    return "`" + s.replace("`", "") + "`"

def q_literal(s: str) -> str:
    return s.replace("'", "''")

def split_db_table(full: str) -> tuple[str, str]:
    db, tbl = full.split(".", 1)
    return db, tbl

def detectar_patron_horas(cur, table_or: str) -> str:
    db, tbl = split_db_table(table_or)

    cur.execute("""
        SELECT COUNT(*)
        FROM INFORMATION_SCHEMA.COLUMNS
        WHERE TABLE_SCHEMA=%s AND TABLE_NAME=%s AND COLUMN_NAME=%s
    """, (db, tbl, "Hora 1"))
    if cur.fetchone()[0] > 0:
        return "Hora {h}"

    cur.execute("""
        SELECT COUNT(*)
        FROM INFORMATION_SCHEMA.COLUMNS
        WHERE TABLE_SCHEMA=%s AND TABLE_NAME=%s AND COLUMN_NAME=%s
    """, (db, tbl, "Hora1"))
    if cur.fetchone()[0] > 0:
        return "Hora{h}"

    raise RuntimeError("No encontré columnas 'Hora 1' ni 'Hora1' en la tabla origen.")

def obtener_subtipos(cur, table_or: str) -> list[str]:
    cur.execute(f"""
        SELECT DISTINCT Subtipo
        FROM {table_or}
        WHERE Subtipo IS NOT NULL
          AND TRIM(Subtipo) <> ''
          AND TRIM(Categoria) <> 'Sistemas de Almacenamiento'
        ORDER BY Subtipo
    """)
    subtipos = [r[0] for r in cur.fetchall()]

    # Agregar columnas especiales para almacenamiento
    for extra in ["Inyección", "Retiro"]:
        if extra not in subtipos:
            subtipos.append(extra)

    return subtipos

def build_unpivot_union(table_or: str, n_horas: int, hora_pattern: str) -> str:
    parts = []
    for h in range(1, n_horas + 1):
        hora_out = h - 1
        col_h = hora_pattern.format(h=h)
        col_h_sql = q_ident(col_h)

        parts.append(f"""
        SELECT
          Fecha,
          {hora_out} AS hora,
          CASE
            WHEN TRIM(Categoria) = 'Sistemas de Almacenamiento' AND COALESCE({col_h_sql}, 0) > 0
              THEN 'Inyección'
            WHEN TRIM(Categoria) = 'Sistemas de Almacenamiento' AND COALESCE({col_h_sql}, 0) < 0
              THEN 'Retiro'
            ELSE Subtipo
          END AS Subtipo_calc,
          CASE
            WHEN TRIM(Categoria) = 'Sistemas de Almacenamiento' AND COALESCE({col_h_sql}, 0) < 0
              THEN COALESCE({col_h_sql}, 0)
            ELSE COALESCE({col_h_sql}, 0)
          END AS valor
        FROM {table_or}
        WHERE Fecha BETWEEN %s AND %s
        """.strip())
    return "\nUNION ALL\n".join(parts)

def ensure_out_table(cur, table_out: str, subtipos: list[str]):
    cols = ",\n  ".join([f"{q_ident(s)} DECIMAL(18,6) NULL" for s in subtipos])

    cur.execute(f"""
    CREATE TABLE IF NOT EXISTS {table_out} (
      Fecha DATE NOT NULL,
      hora  TINYINT NOT NULL,
      {cols},
      PRIMARY KEY (Fecha, hora)
    ) ENGINE=InnoDB;
    """)

    for s in subtipos:
        try:
            cur.execute(f"ALTER TABLE {table_out} ADD COLUMN {q_ident(s)} DECIMAL(18,6) NULL;")
        except mysql.connector.Error:
            pass

def upsert_subtipo_pivot(fecha_inicio, fecha_fin, n_horas: int):
    conn = mysql.connector.connect(
        host=MYSQL_HOST,
        port=MYSQL_PORT,
        user=MYSQL_USER,
        password=MYSQL_PASSWORD,
        database=MYSQL_DB,
    )
    try:
        cur = conn.cursor()

        # 1) Subtipos reales + Inyeccion/Retiro
        subtipos = obtener_subtipos(cur, TABLE_OR)
        print("Subtipos detectados:", len(subtipos))
        print(subtipos)

        # 2) Detectar patrón de horas
        hora_pattern = detectar_patron_horas(cur, TABLE_OR)
        print("Patrón horas detectado:", hora_pattern)

        # 3) Crear/asegurar tabla destino
        ensure_out_table(cur, TABLE_OUT, subtipos)

        # 4) Armar unpivot especial
        unpivot = build_unpivot_union(TABLE_OR, n_horas, hora_pattern)

        cols_insert = ["Fecha", "hora"] + [q_ident(s) for s in subtipos]

        sums = []
        updates = []
        for s in subtipos:
            sums.append(
                f"SUM(CASE WHEN Subtipo_calc = '{q_literal(s)}' THEN valor ELSE 0 END) AS {q_ident(s)}"
            )
            updates.append(
                f"{q_ident(s)} = VALUES({q_ident(s)})"
            )

        sql = f"""
        INSERT INTO {TABLE_OUT} ({", ".join(cols_insert)})
        SELECT
          Fecha,
          hora,
          {", ".join(sums)}
        FROM (
          {unpivot}
        ) u
        WHERE Subtipo_calc IS NOT NULL
          AND TRIM(Subtipo_calc) <> ''
        GROUP BY Fecha, hora
        ON DUPLICATE KEY UPDATE
          {", ".join(updates)};
        """

        params = []
        for _ in range(n_horas):
            params.extend([fecha_inicio, fecha_fin])

        cur.execute(sql, tuple(params))
        conn.commit()

        cur.execute(
            f"SELECT COUNT(*) FROM {TABLE_OUT} WHERE Fecha BETWEEN %s AND %s;",
            (fecha_inicio, fecha_fin)
        )
        print("Filas en tabla destino (rango):", cur.fetchone()[0])

    finally:
        conn.close()

upsert_subtipo_pivot(FECHA_INICIO, FECHA_FIN, N_HORAS)
```

```python
import pandas as pd
import mysql.connector
from datetime import date

# -----------------------
# Config
# -----------------------
MYSQL_HOST = "127.0.0.1"
MYSQL_PORT = 3306
MYSQL_USER = "root"
MYSQL_PASSWORD = "rootvale123"
MYSQL_DB = "costos_variables"

TABLE_OR = "costos_variables.prg_energia_horaria"
 

def parse_lista(texto: str) -> list[str]:
    items = [x.strip() for x in texto.replace("\u00A0", " ").split() if x.strip()]
    return items


CENTRALES = CENTRALES_HIDRO + CENTRALES_TER 

print("Total centrales objetivo:", len(CENTRALES))

def split_db_table(fully_qualified: str) -> tuple[str, str]:
    db, tbl = fully_qualified.split(".", 1)
    return db, tbl

def detectar_patron_horas(cur, table_or: str) -> str:
    db, tbl = split_db_table(table_or)

    cur.execute("""
        SELECT COUNT(*)
        FROM INFORMATION_SCHEMA.COLUMNS
        WHERE TABLE_SCHEMA=%s AND TABLE_NAME=%s AND COLUMN_NAME=%s
    """, (db, tbl, "Hora 1"))
    if cur.fetchone()[0] > 0:
        return "Hora {h}"

    cur.execute("""
        SELECT COUNT(*)
        FROM INFORMATION_SCHEMA.COLUMNS
        WHERE TABLE_SCHEMA=%s AND TABLE_NAME=%s AND COLUMN_NAME=%s
    """, (db, tbl, "Hora1"))
    if cur.fetchone()[0] > 0:
        return "Hora{h}"

    raise RuntimeError("No encontré columnas 'Hora 1' ni 'Hora1' en la tabla origen.")

def _to_date_str(d: date) -> str:
    return d.isoformat()

def build_centrales_df(fecha_inicio: date, fecha_fin: date, n_horas: int = 24) -> pd.DataFrame:
    """
    Devuelve un DF ancho:
      columnas: Fecha, hora, <CENTRALES_BASE...>, <columna_blanca>, <CENTRALES_EXTRA...>
    tomando datos desde costos_variables.prg_energia_horaria.
    No crea tablas en SQL.
    """

    conn = mysql.connector.connect(
        host=MYSQL_HOST, port=MYSQL_PORT,
        user=MYSQL_USER, password=MYSQL_PASSWORD,
        database=MYSQL_DB,
    )

    try:
        cur = conn.cursor()
        hora_pattern = detectar_patron_horas(cur, TABLE_OR)

        hora_cols = [hora_pattern.format(h=h) for h in range(1, n_horas + 1)]
        cols_sql = ", ".join([f"`{c}`" for c in ["Fecha", "Central", *hora_cols]])

        placeholders = ", ".join(["%s"] * len(CENTRALES))
        sql = f"""
            SELECT {cols_sql}
            FROM {TABLE_OR}
            WHERE Fecha BETWEEN %s AND %s
              AND Central IN ({placeholders})
        """

        params = [_to_date_str(fecha_inicio), _to_date_str(fecha_fin), *CENTRALES]
        src = pd.read_sql(sql, conn, params=params)

        if src.empty:
            out_cols = ["Fecha", "hora", *CENTRALES, ""]
            return pd.DataFrame(columns=out_cols)

        # Unpivot
        long = src.melt(
            id_vars=["Fecha", "Central"],
            value_vars=hora_cols,
            var_name="hora_col",
            value_name="valor"
        )

        long["hora"] = (
            long["hora_col"]
            .str.replace("Hora", "", regex=False)
            .str.replace(" ", "", regex=False)
            .astype(int) - 1
        )
        long.drop(columns=["hora_col"], inplace=True)

        # Pivot
        wide = (
            long.pivot_table(
                index=["Fecha", "hora"],
                columns="Central",
                values="valor",
                aggfunc="sum"
            )
            .reset_index()
        )

        # Asegurar columnas en el orden pedido
        for c in CENTRALES:
            if c not in wide.columns:
                wide[c] = pd.NA

        wide[""] = pd.NA

        # Orden final
        wide = wide[["Fecha", "hora", *CENTRALES, ""]]

        # Tipos
        wide["Fecha"] = pd.to_datetime(wide["Fecha"]).dt.date
        wide["hora"] = wide["hora"].astype(int)

        return wide

    finally:
        conn.close()


centrales_df = build_centrales_df(FECHA_INICIO, FECHA_FIN, N_HORAS)

centrales_df.to_pickle("centrales_prg_energia_horaria.pkl")

print("Orden columnas:")
print(list(centrales_df.columns))
```

```python
import pandas as pd
import mysql.connector
from datetime import date

MYSQL = dict(
    host="127.0.0.1",
    port=3306,
    user="root",
    password="rootvale123",
    database="costos_variables",
)

TABLE_RD = "costos_variables.resumen_diario_central"   # Fecha, Central, P_ponderado (diario)
TABLE_PO = "costos_variables.po"                       # Fecha, Central, Combustible_Costo (diario)

def read_diario(conn, table_src: str, col_precio: str,
                fecha_inicio, fecha_fin, centrales: list[str]) -> pd.DataFrame:
    """
    Lee (Fecha, Central, col_precio) desde MySQL filtrando por fechas y centrales.
    Devuelve DF largo (diario).
    """
    placeholders = ", ".join(["%s"] * len(centrales))
    sql = f"""
        SELECT Fecha, Central, {col_precio} AS precio
        FROM {table_src}
        WHERE Fecha BETWEEN %s AND %s
          AND Central IN ({placeholders})
    """
    params = [fecha_inicio, fecha_fin, *centrales]
    return pd.read_sql(sql, conn, params=params)

def diario_a_horas_wide(df_diario: pd.DataFrame, centrales: list[str], n_horas: int = 24) -> pd.DataFrame:
    """
    Expande diario -> horas (0..n_horas-1) y pivotea a ancho:
      index: Fecha, hora
      columns: centrales
      values: precio
    """
    df = df_diario.copy()
    # normaliza columnas
    if "Central" not in df.columns:
        raise ValueError("df_diario debe traer columna 'Central'")
    if "Fecha" not in df.columns:
        raise ValueError("df_diario debe traer columna 'Fecha'")
    if "precio" not in df.columns:
        raise ValueError("df_diario debe traer columna 'precio'")

    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.date
    df["Central"] = df["Central"].astype(str)

    # cross join con horas
    horas = pd.DataFrame({"hora": list(range(n_horas))}, dtype="int16")
    df["__k"] = 1
    horas["__k"] = 1
    dfh = df.merge(horas, on="__k", how="inner").drop(columns="__k")

    # pivot ancho
    wide = (
        dfh.pivot_table(
            index=["Fecha", "hora"],
            columns="Central",
            values="precio",
            aggfunc="max",
        )
        .reset_index()
    )

    # asegura orden/columnas completas según CENTRALES
    for c in centrales:
        if c not in wide.columns:
            wide[c] = pd.NA
    wide = wide[["Fecha", "hora"] + list(centrales)]

    return wide

def build_precio_comb_pandas(fecha_inicio, fecha_fin, centrales):
    with mysql.connector.connect(**MYSQL) as conn:

        df_po = read_diario(conn, TABLE_PO, "Combustible_Costo", fecha_inicio, fecha_fin, centrales)
        precio_po = diario_a_horas_wide(df_po, centrales, N_HORAS)

    return precio_po

precio_po_df = build_precio_comb_pandas(FECHA_INICIO, FECHA_FIN, CENTRALES_TER)



precio_po_df.to_pickle("precio_combustible_po.pkl")
```

```python
import pandas as pd
import mysql.connector

# ----------------------------
# Utils comunes
# ----------------------------
def norm(s: str) -> str:
    s = str(s).replace("\u00A0", " ").strip().upper()
    s = " ".join(s.split())
    return s

def qident(col: str) -> str:
    return f"`{col.replace('`','``')}`"

def mysql_connect(MYSQL: dict):
    return mysql.connector.connect(**MYSQL)

def read_mapping_homologa(conn,
                          table_map: str,
                          col_src: str,
                          col_dst: str,
                          src_alias: str = "Central",
                          dst_alias: str = "Homologo") -> pd.DataFrame:
    """
    Lee mapeo genérico desde una tabla homóloga.
    Ejemplo:
      - PO/TCO Central -> PRG Central
      - PO/TCO Central -> otra clave equivalente
    """
    sql = f"""
        SELECT
          TRIM({qident(col_src)}) AS {qident(src_alias)},
          TRIM({qident(col_dst)}) AS {qident(dst_alias)}
        FROM {table_map}
        WHERE {qident(col_src)} IS NOT NULL AND TRIM({qident(col_src)}) <> ''
          AND {qident(col_dst)} IS NOT NULL AND TRIM({qident(col_dst)}) <> ''
    """
    df = pd.read_sql(sql, conn)
    df[src_alias] = df[src_alias].astype(str).map(norm)
    df[dst_alias] = df[dst_alias].astype(str).map(norm)
    df = df.drop_duplicates(subset=[src_alias, dst_alias])
    return df

def build_unpivot_union(table_name: str,
                        fecha_col: str,
                        entity_col: str,
                        hour_cols: list[str],
                        entity_alias: str = "Central",
                        value_alias: str = "valor") -> str:
    """
    SQL genérico para unpivotear columnas horarias a filas:
      Fecha, hora, <entity_alias>, <value_alias>
    """
    parts = []
    for i, hc in enumerate(hour_cols, start=1):
        hora_out = i - 1
        parts.append(f"""
        SELECT
            {qident(fecha_col)} AS Fecha,
            {hora_out} AS hora,
            TRIM({qident(entity_col)}) AS {qident(entity_alias)},
            COALESCE({qident(hc)}, 0) AS {qident(value_alias)}
        FROM {table_name}
        WHERE {qident(fecha_col)} BETWEEN %s AND %s
          AND {qident(entity_col)} IS NOT NULL
          AND TRIM({qident(entity_col)}) <> ''
        """.strip())
    return "\nUNION ALL\n".join(parts)

def read_hourly_long(conn,
                     table_name: str,
                     fecha_inicio,
                     fecha_fin,
                     fecha_col: str = "Fecha",
                     entity_col: str = "Central",
                     hour_cols: list[str] | None = None,
                     entity_alias: str = "Central",
                     value_alias: str = "valor") -> pd.DataFrame:
    """
    Lector genérico de tablas horarias wide -> long.
    Retorna:
      Fecha, hora, <entity_alias>, <value_alias>
    """
    if hour_cols is None:
        hour_cols = [f"Hora{i}" for i in range(1, 25)]

    unpivot_sql = build_unpivot_union(
        table_name=table_name,
        fecha_col=fecha_col,
        entity_col=entity_col,
        hour_cols=hour_cols,
        entity_alias=entity_alias,
        value_alias=value_alias,
    )

    params = []
    for _ in range(len(hour_cols)):
        params.extend([fecha_inicio, fecha_fin])

    sql = f"""
    SELECT
      Fecha,
      hora,
      {qident(entity_alias)},
      SUM({qident(value_alias)}) AS {qident(value_alias)}
    FROM (
      {unpivot_sql}
    ) x
    GROUP BY Fecha, hora, {qident(entity_alias)}
    """

    df = pd.read_sql(sql, conn, params=params)
    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.date
    df["hora"] = pd.to_numeric(df["hora"], errors="coerce").astype("Int64")
    df[entity_alias] = df[entity_alias].astype(str).map(norm)
    df[value_alias] = pd.to_numeric(df[value_alias], errors="coerce")
    return df

def read_daily_value_subtipo(conn,
                             table_name: str,
                             fecha_inicio,
                             fecha_fin,
                             fecha_col: str = "Fecha",
                             central_col: str = "Central",
                             value_col: str = "valor",
                             subtipo_col: str = "Subtipo",
                             value_alias: str = "valor",
                             agg_value: str = "mean") -> pd.DataFrame:
    """
    Lee una tabla diaria de valores por central y subtipo.
    Retorna:
      Fecha, Central, <value_alias>, Subtipo
    """
    sql = f"""
        SELECT
          {qident(fecha_col)} AS Fecha,
          TRIM({qident(central_col)}) AS Central,
          {qident(value_col)} AS {qident(value_alias)},
          TRIM({qident(subtipo_col)}) AS Subtipo
        FROM {table_name}
        WHERE {qident(fecha_col)} BETWEEN %s AND %s
          AND {qident(central_col)} IS NOT NULL
          AND TRIM({qident(central_col)}) <> ''
    """
    df = pd.read_sql(sql, conn, params=[fecha_inicio, fecha_fin])
    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.date
    df["Central"] = df["Central"].astype(str).map(norm)
    df[value_alias] = pd.to_numeric(df[value_alias], errors="coerce")
    df["Subtipo"] = df["Subtipo"].astype(str).map(norm)

    df = (
        df.groupby(["Fecha", "Central"], as_index=False)
          .agg(
              **{
                  value_alias: (value_alias, agg_value),
                  "Subtipo": ("Subtipo", "first"),
              }
          )
    )
    return df

def weighted_min_max_by_group(df: pd.DataFrame,
                              group_cols: list[str],
                              value_col: str,
                              weight_col: str,
                              out_prefix: str = "VAL") -> pd.DataFrame:
    """
    Calcula por grupo:
      - ponderado = sum(value*weight)/sum(weight)
      - min(value)
      - max(value)
    """
    d = df.copy()
    d["_num"] = d[value_col] * d[weight_col]

    g = d.groupby(group_cols, as_index=False).agg(
        num=("_num", "sum"),
        den=(weight_col, "sum"),
        vmin=(value_col, "min"),
        vmax=(value_col, "max"),
    )
    g[f"{out_prefix}_PON"] = g["num"] / g["den"]
    g.loc[g["den"] == 0, f"{out_prefix}_PON"] = pd.NA

    g = g.drop(columns=["num", "den"])
    g = g.rename(columns={"vmin": f"{out_prefix}_MIN", "vmax": f"{out_prefix}_MAX"})
    return g

def pivot_three_metrics(out_long: pd.DataFrame,
                        index_cols: list[str],
                        subtipo_col: str,
                        prefix: str) -> pd.DataFrame:
    """
    out_long debe traer columnas:
      - index_cols
      - subtipo_col
      - f"{prefix}_PON", f"{prefix}_MIN", f"{prefix}_MAX"
    """
    out_long = out_long.copy()
    out_long[subtipo_col] = out_long[subtipo_col].fillna("SIN_SUBTIPO")

    wide_parts = []
    for suf in ["PON", "MIN", "MAX"]:
        valcol = f"{prefix}_{suf}"
        w = (
            out_long.pivot_table(
                index=index_cols,
                columns=subtipo_col,
                values=valcol,
                aggfunc="first",
            )
            .reset_index()
        )
        rename = {c: f"{c}_{suf}" for c in w.columns if c not in index_cols}
        w = w.rename(columns=rename)
        wide_parts.append(w)

    wide = wide_parts[0]
    for w in wide_parts[1:]:
        wide = wide.merge(w, on=index_cols, how="outer")

    wide = wide.sort_values(index_cols).reset_index(drop=True)
    return wide
```

```python
import pandas as pd
import mysql.connector
from datetime import date

# ----------------------------
# CONFIG
# ----------------------------
MYSQL = dict(
    host="127.0.0.1",
    port=3306,
    user="root",
    password="rootvale123",
    database="costos_variables",
)

TABLE_PO   = "costos_variables.po"
TABLE_GEN  = "costos_variables.prg_energia_horaria"
TABLE_MAP  = "mapeo.homologa"

COL_FECHA   = "Fecha"
COL_CENTRAL = "Central"
COL_PRECIO  = "Combustible_Costo"

MAP_COL_CLAVE = "clave_po"
MAP_COL_SUB   = "Sub"

GEN_CENTRAL_COL = "Central"
GEN_HOUR_COLS = [f"Hora{i}" for i in range(1, 25)]

GEN_EPS = 0.0

SUBTIPOS_PONDERADOS = {"COGENERACION", "CARBÓN", "GNA", "GNL","BIOMASA"}


# ----------------------------
# UTILS
# ----------------------------
def norm(s: str) -> str:
    s = str(s).replace("\u00A0", " ").strip().upper()
    s = " ".join(s.split())
    return s

def qident(col: str) -> str:
    return f"`{col.replace('`','``')}`"

def build_gen_unpivot_union(table_gen: str, hour_cols: list[str]) -> str:
    parts = []
    for i, hc in enumerate(hour_cols, start=1):
        hora_out = i - 1
        parts.append(f"""
        SELECT
            {qident(COL_FECHA)} AS Fecha,
            {hora_out} AS hora,
            TRIM({qident(GEN_CENTRAL_COL)}) AS Central,
            COALESCE({qident(hc)}, 0) AS gen
        FROM {table_gen}
        WHERE {qident(COL_FECHA)} BETWEEN %s AND %s
          AND {qident(GEN_CENTRAL_COL)} IS NOT NULL
          AND TRIM({qident(GEN_CENTRAL_COL)}) <> ''
        """.strip())
    return "\nUNION ALL\n".join(parts)


# ----------------------------
# READS
# ----------------------------
def read_homologa_subtipo(conn) -> pd.DataFrame:
    sql = f"""
        SELECT
          TRIM({qident(MAP_COL_CLAVE)}) AS Central,
          TRIM({qident(MAP_COL_SUB)}) AS Subtipo
        FROM {TABLE_MAP}
        WHERE {qident(MAP_COL_CLAVE)} IS NOT NULL
          AND TRIM({qident(MAP_COL_CLAVE)}) <> ''
          AND {qident(MAP_COL_SUB)} IS NOT NULL
          AND TRIM({qident(MAP_COL_SUB)}) <> ''
    """
    df = pd.read_sql(sql, conn)
    df["Central"] = df["Central"].astype(str).map(norm)
    df["Subtipo"] = df["Subtipo"].astype(str).map(norm)
    df = df.drop_duplicates(subset=["Central", "Subtipo"])
    return df

def read_po_precio(conn, fecha_inicio, fecha_fin) -> pd.DataFrame:
    sql = f"""
        SELECT
          {qident(COL_FECHA)} AS Fecha,
          TRIM({qident(COL_CENTRAL)}) AS Central,
          MAX({qident(COL_PRECIO)}) AS precio
        FROM {TABLE_PO}
        WHERE {qident(COL_FECHA)} BETWEEN %s AND %s
          AND {qident(COL_CENTRAL)} IS NOT NULL
          AND TRIM({qident(COL_CENTRAL)}) <> ''
        GROUP BY {qident(COL_FECHA)}, TRIM({qident(COL_CENTRAL)})
    """
    df = pd.read_sql(sql, conn, params=[fecha_inicio, fecha_fin])
    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.date
    df["Central"] = df["Central"].astype(str).map(norm)
    df["precio"] = pd.to_numeric(df["precio"], errors="coerce")
    return df

def read_gen_horaria_long(conn, fecha_inicio, fecha_fin) -> pd.DataFrame:
    unpivot_sql = build_gen_unpivot_union(TABLE_GEN, GEN_HOUR_COLS)
    params = []
    for _ in range(len(GEN_HOUR_COLS)):
        params.extend([fecha_inicio, fecha_fin])

    sql = f"""
    SELECT
      Fecha,
      hora,
      Central,
      SUM(gen) AS gen
    FROM (
      {unpivot_sql}
    ) x
    GROUP BY Fecha, hora, Central
    """
    df = pd.read_sql(sql, conn, params=params)
    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.date
    df["hora"] = pd.to_numeric(df["hora"], errors="coerce").astype("Int64")
    df["Central"] = df["Central"].astype(str).map(norm)
    df["gen"] = pd.to_numeric(df["gen"], errors="coerce")
    return df


# ----------------------------
# MAIN
# ----------------------------
def combustible_pond_minmax_por_subtipo_horario(fecha_inicio, fecha_fin):
    with mysql.connector.connect(**MYSQL) as conn:
        po   = read_po_precio(conn, fecha_inicio, fecha_fin)
        gen  = read_gen_horaria_long(conn, fecha_inicio, fecha_fin)
        homo = read_homologa_subtipo(conn)

    # precio + subtipo desde homologa
    po = po.merge(homo, on="Central", how="left")

    # base con generación + precio/subtipo
    base = gen.merge(po, on=["Fecha", "Central"], how="left")

    centrales_gen_sin_po = sorted(
        base.loc[base["precio"].isna(), "Central"].dropna().drop_duplicates().tolist()
    )

    centrales_sin_sub_homologa = sorted(
        base.loc[base["Subtipo"].isna(), "Central"].dropna().drop_duplicates().tolist()
    )

    # =========
    # PARTE 1: SUBTIPOS PONDERADOS
    # =========
    calc_pond = base.dropna(subset=["gen", "precio", "Subtipo"]).copy()
    calc_pond = calc_pond.loc[calc_pond["gen"] > GEN_EPS].copy()
    calc_pond = calc_pond.loc[calc_pond["Subtipo"].isin(SUBTIPOS_PONDERADOS)].copy()

    if not calc_pond.empty:
        calc_pond["num"] = calc_pond["precio"] * calc_pond["gen"]

        out_pond = (
            calc_pond.groupby(["Fecha", "hora", "Subtipo"], as_index=False)
                .agg(
                    num=("num", "sum"),
                    den=("gen", "sum"),
                    COMB_MIN=("precio", "min"),
                    COMB_MAX=("precio", "max"),
                    n_centrales=("Central", "nunique"),
                )
        )
        out_pond["COMB_PON"] = out_pond["num"] / out_pond["den"]
        out_pond.loc[out_pond["den"] == 0, ["COMB_PON", "COMB_MIN", "COMB_MAX"]] = pd.NA
        out_pond = out_pond.rename(columns={"den": "GEN_MWh"}).drop(columns=["num"])
    else:
        out_pond = pd.DataFrame(columns=["Fecha", "hora", "Subtipo", "COMB_PON", "COMB_MIN", "COMB_MAX", "GEN_MWh", "n_centrales"])

    # =========
    # PARTE 2: SUBTIPOS SIMPLES
    # =========
    po_simple = po.dropna(subset=["precio", "Subtipo"]).copy()
    po_simple = po_simple.loc[~po_simple["Subtipo"].isin(SUBTIPOS_PONDERADOS)].copy()

    if not po_simple.empty:
        po_simple_daily = (
            po_simple.groupby(["Fecha", "Subtipo"], as_index=False)
                .agg(
                    COMB_PON=("precio", "mean"),
                    COMB_MIN=("precio", "min"),
                    COMB_MAX=("precio", "max"),
                    n_centrales=("Central", "nunique"),
                )
        )
    else:
        po_simple_daily = pd.DataFrame(columns=["Fecha", "Subtipo", "COMB_PON", "COMB_MIN", "COMB_MAX", "n_centrales"])

    base_simple_gen = base.dropna(subset=["gen", "Subtipo"]).copy()
    base_simple_gen = base_simple_gen.loc[base_simple_gen["gen"] > GEN_EPS].copy()
    base_simple_gen = base_simple_gen.loc[~base_simple_gen["Subtipo"].isin(SUBTIPOS_PONDERADOS)].copy()

    if not base_simple_gen.empty:
        gen_simple_h = (
            base_simple_gen.groupby(["Fecha", "hora", "Subtipo"], as_index=False)
                .agg(GEN_MWh=("gen", "sum"))
        )
    else:
        gen_simple_h = pd.DataFrame(columns=["Fecha", "hora", "Subtipo", "GEN_MWh"])

    horas = gen[["Fecha", "hora"]].drop_duplicates().copy()
    if not po_simple_daily.empty and not horas.empty:
        horas["_k"] = 1
        po_simple_daily["_k"] = 1
        out_simple = horas.merge(po_simple_daily, on=["Fecha", "_k"], how="inner").drop(columns="_k")
        out_simple = out_simple.merge(gen_simple_h, on=["Fecha", "hora", "Subtipo"], how="left")
        out_simple["GEN_MWh"] = out_simple["GEN_MWh"].fillna(0.0)
    else:
        out_simple = pd.DataFrame(columns=["Fecha", "hora", "Subtipo", "COMB_PON", "COMB_MIN", "COMB_MAX", "n_centrales", "GEN_MWh"])

    out_long = pd.concat([out_pond, out_simple], ignore_index=True)
    out_long = out_long.sort_values(["Fecha", "hora", "Subtipo"]).reset_index(drop=True)

    if out_long.empty:
        out_wide = pd.DataFrame(columns=["Fecha", "hora"])
        diag = dict(
            total_centrales_po=po["Central"].nunique(),
            total_centrales_gen=gen["Central"].nunique(),
            centrales_gen_sin_po=len(centrales_gen_sin_po),
            lista_gen_sin_po=centrales_gen_sin_po,
            centrales_sin_sub_homologa=len(centrales_sin_sub_homologa),
            lista_sin_sub_homologa=centrales_sin_sub_homologa,
        )
        return out_wide, out_long, base, diag

    out_long["Subtipo"] = out_long["Subtipo"].fillna("SIN_SUBTIPO")

    wide_pon = out_long.pivot_table(
        index=["Fecha", "hora"], columns="Subtipo", values="COMB_PON", aggfunc="first"
    ).reset_index()

    wide_min = out_long.pivot_table(
        index=["Fecha", "hora"], columns="Subtipo", values="COMB_MIN", aggfunc="first"
    ).reset_index()

    wide_max = out_long.pivot_table(
        index=["Fecha", "hora"], columns="Subtipo", values="COMB_MAX", aggfunc="first"
    ).reset_index()

    wide_pon = wide_pon.rename(columns={c: f"{c}_PON" for c in wide_pon.columns if c not in ["Fecha", "hora"]})
    wide_min = wide_min.rename(columns={c: f"{c}_MIN" for c in wide_min.columns if c not in ["Fecha", "hora"]})
    wide_max = wide_max.rename(columns={c: f"{c}_MAX" for c in wide_max.columns if c not in ["Fecha", "hora"]})

    out_wide = (
        wide_pon.merge(wide_min, on=["Fecha", "hora"], how="outer")
                .merge(wide_max, on=["Fecha", "hora"], how="outer")
                .sort_values(["Fecha", "hora"])
                .reset_index(drop=True)
    )

    diag = dict(
        total_centrales_po=po["Central"].nunique(),
        total_centrales_gen=gen["Central"].nunique(),
        centrales_gen_sin_po=len(centrales_gen_sin_po),
        lista_gen_sin_po=centrales_gen_sin_po,
        centrales_sin_sub_homologa=len(centrales_sin_sub_homologa),
        lista_sin_sub_homologa=centrales_sin_sub_homologa,
    )

    return out_wide, out_long, base, diag

out_wide_df, out_long_df, base_df, diag = combustible_pond_minmax_por_subtipo_horario(FECHA_INICIO, FECHA_FIN)

print("Diagnóstico:")
print(f"- Centrales únicas en PO (rango): {diag['total_centrales_po']}")
print(f"- Centrales únicas en PRG (rango): {diag['total_centrales_gen']}")
print(f"- Centrales PRG sin precio PO: {diag['centrales_gen_sin_po']}")
print(f"- Centrales sin subtipo en homologa: {diag['centrales_sin_sub_homologa']}")
```

### Cargar en SQL

```python
import numpy as np
import pandas as pd
import mysql.connector

TABLE_OUT_SQL = "costos_variables.combustible_pond_minmax_subtipo_horario"
def mysql_connect(mysql_cfg: dict):
    return mysql.connector.connect(**mysql_cfg)
def qi(name: str) -> str:
    return f"`{str(name).replace('`','``')}`"

def qtable(full: str) -> str:
    return ".".join(qi(p) for p in full.split("."))

def ensure_table_wide(cur, table_out: str, value_cols: list[str]):
    tbl = qtable(table_out)
    cur.execute(f"""
        CREATE TABLE IF NOT EXISTS {tbl} (
          Fecha DATE NOT NULL,
          hora  TINYINT NOT NULL,
          PRIMARY KEY (Fecha, hora)
        ) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4 COLLATE=utf8mb4_0900_ai_ci;
    """)
    schema, tname = table_out.split(".", 1)
    cur.execute(f"SHOW COLUMNS FROM {qi(schema)}.{qi(tname)};")
    existing = {row[0] for row in cur.fetchall()}
    for c in value_cols:
        if c in existing:
            continue
        cur.execute(f"ALTER TABLE {tbl} ADD COLUMN {qi(c)} DECIMAL(18,6) NULL;")

def df_to_mysql_none(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy().astype(object)
    out.replace([np.nan, np.inf, -np.inf], None, inplace=True)
    return out

def save_wide_to_mysql(out_wide_df: pd.DataFrame, table_out: str, batch_size: int = 500):
    if out_wide_df.empty:
        print("DF vacío: no se escribió nada.")
        return

    df = out_wide_df.copy()
    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.date
    df["hora"] = pd.to_numeric(df["hora"], errors="coerce").astype(int)

    key_cols = ["Fecha", "hora"]
    value_cols = [c for c in df.columns if c not in key_cols]
    insert_cols = key_cols + value_cols

    df = df[insert_cols]
    df = df_to_mysql_none(df)

    tbl = qtable(table_out)
    cols_insert_sql = ", ".join(qi(c) for c in insert_cols)
    placeholders = ", ".join(["%s"] * len(insert_cols))
    updates = ", ".join([f"{qi(c)}=VALUES({qi(c)})" for c in value_cols])

    sql = f"""
        INSERT INTO {tbl} ({cols_insert_sql})
        VALUES ({placeholders})
        ON DUPLICATE KEY UPDATE {updates};
    """

    with mysql_connect(MYSQL) as conn:
        cur = conn.cursor()
        cur.execute("SET SESSION net_read_timeout=3600")
        cur.execute("SET SESSION net_write_timeout=3600")
        cur.execute("SET SESSION wait_timeout=28800")
        cur.execute("SET SESSION interactive_timeout=28800")

        ensure_table_wide(cur, table_out, value_cols)
        conn.commit()

        data = df.to_numpy(dtype=object)
        nan_left = sum(1 for x in data.ravel() if isinstance(x, float) and x != x)
        if nan_left:
            raise RuntimeError(f"Quedaron {nan_left} NaN float en el array final; no se puede insertar.")

        total = len(data)
        for i in range(0, total, batch_size):
            cur.executemany(sql, data[i:i+batch_size].tolist())
            conn.commit()
            print(f"Upsert: {min(i+batch_size, total)}/{total}")

        cur.execute(f"SELECT COUNT(*) FROM {tbl};")
        print("Total filas en tabla destino:", cur.fetchone()[0])

# Guardar
save_wide_to_mysql(out_wide_df, TABLE_OUT_SQL, batch_size=500)
```

```python
import pandas as pd
import mysql.connector
from datetime import date

# ----------------------------
# CONFIG
# ----------------------------
MYSQL = dict(
    host="127.0.0.1",
    port=3306,
    user="root",
    password="rootvale123",
    database="costos_variables",
)

TABLE_PO   = "costos_variables.po"
TABLE_GEN  = "costos_variables.prg_energia_horaria"
TABLE_MAP  = "mapeo.homologa"

COL_FECHA   = "Fecha"
COL_CENTRAL = "Central"
COL_PRECIO  = "Combustible_Costo"
COL_SUBTIPO = "Subtipo"

MAP_COL_CLAVE_PO = "clave_po"
MAP_COL_EMPRESA  = "Empresa 2"

GEN_CENTRAL_COL = "Central"
GEN_HOUR_COLS = [f"Hora{i}" for i in range(1, 25)]


EMPRESAS_OBJETIVO = [
    "COLBÚN",
    "ENEL SIC",
    "ENEL SING",
    "GMETROPOLITANA",
    "ENGIE",
    "TAMAKAYA",
]

SUBTIPOS_EMPRESA = ["GNA", "GNL"]

def norm(s: str) -> str:
    s = str(s).replace("\u00A0", " ").strip().upper()
    s = " ".join(s.split())
    return s

def qident(col: str) -> str:
    return f"`{col.replace('`','``')}`"

def build_gen_unpivot_union(table_gen: str, hour_cols: list[str]) -> str:
    parts = []
    for i, hc in enumerate(hour_cols, start=1):
        hora_out = i - 1
        parts.append(f"""
        SELECT
            {qident(COL_FECHA)} AS Fecha,
            {hora_out} AS hora,
            TRIM({qident(GEN_CENTRAL_COL)}) AS Central,
            COALESCE({qident(hc)}, 0) AS gen
        FROM {table_gen}
        WHERE {qident(COL_FECHA)} BETWEEN %s AND %s
          AND {qident(GEN_CENTRAL_COL)} IS NOT NULL
          AND TRIM({qident(GEN_CENTRAL_COL)}) <> ''
        """.strip())
    return "\nUNION ALL\n".join(parts)

def read_po_precio_subtipo(conn, fecha_inicio, fecha_fin) -> pd.DataFrame:
    sql = f"""
        SELECT
          {qident(COL_FECHA)} AS Fecha,
          TRIM({qident(COL_CENTRAL)}) AS Central,
          MAX({qident(COL_PRECIO)}) AS precio,
          MAX(TRIM({qident(COL_SUBTIPO)})) AS Subtipo
        FROM {TABLE_PO}
        WHERE {qident(COL_FECHA)} BETWEEN %s AND %s
          AND {qident(COL_CENTRAL)} IS NOT NULL
          AND TRIM({qident(COL_CENTRAL)}) <> ''
        GROUP BY {qident(COL_FECHA)}, TRIM({qident(COL_CENTRAL)})
    """
    df = pd.read_sql(sql, conn, params=[fecha_inicio, fecha_fin])
    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.date
    df["Central"] = df["Central"].astype(str).map(norm)
    df["precio"] = pd.to_numeric(df["precio"], errors="coerce")
    df["Subtipo"] = df["Subtipo"].astype(str).map(norm)
    return df

def read_gen_horaria_long(conn, fecha_inicio, fecha_fin) -> pd.DataFrame:
    unpivot_sql = build_gen_unpivot_union(TABLE_GEN, GEN_HOUR_COLS)
    params = []
    for _ in range(len(GEN_HOUR_COLS)):
        params.extend([fecha_inicio, fecha_fin])

    sql = f"""
    SELECT
      Fecha,
      hora,
      Central,
      SUM(gen) AS gen
    FROM (
      {unpivot_sql}
    ) x
    GROUP BY Fecha, hora, Central
    """
    df = pd.read_sql(sql, conn, params=params)
    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.date
    df["hora"] = pd.to_numeric(df["hora"], errors="coerce").astype("Int64")
    df["Central"] = df["Central"].astype(str).map(norm)
    df["gen"] = pd.to_numeric(df["gen"], errors="coerce")
    return df

def read_empresas_homologa(conn) -> pd.DataFrame:
    sql = f"""
        SELECT
          TRIM({qident(MAP_COL_CLAVE_PO)}) AS Central,
          TRIM({qident(MAP_COL_EMPRESA)}) AS Empresa
        FROM {TABLE_MAP}
        WHERE {qident(MAP_COL_CLAVE_PO)} IS NOT NULL
          AND TRIM({qident(MAP_COL_CLAVE_PO)}) <> ''
          AND {qident(MAP_COL_EMPRESA)} IS NOT NULL
          AND TRIM({qident(MAP_COL_EMPRESA)}) <> ''
    """
    df = pd.read_sql(sql, conn)
    df["Central"] = df["Central"].astype(str).map(norm)
    df["Empresa"] = df["Empresa"].astype(str).map(norm)
    df = df.drop_duplicates(subset=["Central", "Empresa"])
    return df

def combustible_pond_gna_gnl_por_empresa(fecha_inicio, fecha_fin):
    with mysql.connector.connect(**MYSQL) as conn:
        po = read_po_precio_subtipo(conn, fecha_inicio, fecha_fin)
        gen = read_gen_horaria_long(conn, fecha_inicio, fecha_fin)
        emp = read_empresas_homologa(conn)

    # unir empresa al PO
    po_emp = po.merge(emp, on="Central", how="left")

    # solo GNA/GNL y empresas objetivo
    po_emp = po_emp.loc[po_emp["Subtipo"].isin(SUBTIPOS_EMPRESA)].copy()
    po_emp = po_emp.loc[po_emp["Empresa"].isin(EMPRESAS_OBJETIVO)].copy()

    # join con generación
    base = gen.merge(po_emp, on=["Fecha", "Central"], how="left")

    # solo donde hay gen > 0 y datos válidos
    calc = base.dropna(subset=["gen", "precio", "Subtipo", "Empresa"]).copy()
    calc = calc.loc[calc["gen"] > 0].copy()

    if calc.empty:
        return pd.DataFrame(columns=["Fecha", "hora"])

    calc["num"] = calc["precio"] * calc["gen"]

    out_long = (
        calc.groupby(["Fecha", "hora", "Subtipo", "Empresa"], as_index=False)
            .agg(
                num=("num", "sum"),
                den=("gen", "sum"),
            )
    )

    out_long["COMB_PON"] = out_long["num"] / out_long["den"]
    out_long.loc[out_long["den"] == 0, "COMB_PON"] = pd.NA

    out_long["colname"] = out_long["Subtipo"] + "_" + out_long["Empresa"] + "_PON"

    out_wide = (
        out_long.pivot_table(
            index=["Fecha", "hora"],
            columns="colname",
            values="COMB_PON",
            aggfunc="first"
        )
        .reset_index()
    )

    out_wide.columns.name = None
    out_wide = out_wide.sort_values(["Fecha", "hora"]).reset_index(drop=True)

    return out_wide

out_empresas_df = combustible_pond_gna_gnl_por_empresa(FECHA_INICIO, FECHA_FIN)

print(out_empresas_df.head())
```

```python
import pandas as pd
import mysql.connector

TABLE_TCO = "costos_variables.tco"          # Fecha, Central (nombre TCO), Eficiencia
TABLE_MAP = "mapeo.homologa"     # clave_prg_po, clave_tco


def read_table(conn, table: str, cols: list[str] | None = None, where: str = "", params=None) -> pd.DataFrame:
    sel = "*" if cols is None else ", ".join(cols)
    sql = f"SELECT {sel} FROM {table} " + (f"WHERE {where}" if where else "")
    return pd.read_sql(sql, conn, params=params)

def build_cons_esp_from_tco_pandas(
    fecha_inicio,
    fecha_fin,
    centrales_prg_po: list[str],
    n_horas: int = 24,
    incluir_hora25: bool = False,
) -> pd.DataFrame:
    """
    Construye DF ancho (Fecha, hora, <clave_prg_po...>) usando:
      - tco: (Fecha, Central[TCO], Eficiencia)
      - homologa: mapea clave_tco -> clave_prg_po
    En salida, las columnas se llaman clave_prg_po (las 'CENTRALES' solicitadas).
    Replica valor a todas las horas.
    """
    horas = list(range(n_horas)) + ([24] if incluir_hora25 else [])
    horas_df = pd.DataFrame({"hora": horas}, dtype="int16")

    with mysql.connector.connect(**MYSQL) as conn:
        # 1) mapeo prg_po -> tco (filtrado a las centrales objetivo)
        ph = ", ".join(["%s"] * len(centrales_prg_po))
        map_sql = f"""
            SELECT clave_po, clave_tco
            FROM {TABLE_MAP}
            WHERE clave_po IN ({ph})
              AND clave_tco IS NOT NULL AND clave_tco <> ''
        """
        hom = pd.read_sql(map_sql, conn, params=list(centrales_prg_po))

        if hom.empty:
            raise ValueError("No se encontró mapeo en homologa para las CENTRALES dadas (clave_prg_po).")

        # solo mapeos únicos (por si hay duplicados)
        hom = hom.drop_duplicates(subset=["clave_po", "clave_tco"])

        # 2) traer TCO solo para los clave_tco necesarios y rango fechas
        tco_keys = hom["clave_tco"].dropna().astype(str).unique().tolist()
        ph2 = ", ".join(["%s"] * len(tco_keys))
        tco_sql = f"""
            SELECT Fecha, Central, Eficiencia
            FROM {TABLE_TCO}
            WHERE Fecha BETWEEN %s AND %s
              AND Central IN ({ph2})
        """
        tco = pd.read_sql(tco_sql, conn, params=[fecha_inicio, fecha_fin, *tco_keys])

    # 3) join TCO + homologa para obtener clave_prg_po
    tco["Fecha"] = pd.to_datetime(tco["Fecha"]).dt.date
    tco["Central"] = tco["Central"].astype(str)

    df = tco.merge(
        hom.rename(columns={"clave_po": "central_prg_po", "clave_tco": "central_tco"}),
        how="inner",
        left_on="Central",
        right_on="central_tco",
    )

    # quedarnos con la clave_prg_po y el valor
    df = df[["Fecha", "central_prg_po", "Eficiencia"]].rename(columns={"Eficiencia": "valor"})

    # 4) replicar a horas (cross join)
    df["__k"] = 1
    horas_df["__k"] = 1
    dfh = df.merge(horas_df, on="__k", how="inner").drop(columns="__k")

    # 5) pivot ancho (Fecha, hora) x central_prg_po
    wide = (
        dfh.pivot_table(
            index=["Fecha", "hora"],
            columns="central_prg_po",
            values="valor",
            aggfunc="max",
        )
        .reset_index()
    )

    # 6) asegurar columnas finales en el orden de las centrales solicitadas
    for c in centrales_prg_po:
        if c not in wide.columns:
            wide[c] = pd.NA
    wide = wide[["Fecha", "hora"] + list(centrales_prg_po)]

    return wide

cons_esp_df = build_cons_esp_from_tco_pandas(
    fecha_inicio=FECHA_INICIO,
    fecha_fin=FECHA_FIN,
    centrales_prg_po=CENTRALES_TER,
    n_horas=N_HORAS,
    incluir_hora25=False,
)

cons_esp_df.to_pickle("cons_esp.pkl")
```

```python
import pandas as pd
import mysql.connector
from datetime import date
import numpy as np

MYSQL = dict(
    host="127.0.0.1",
    port=3306,
    user="root",
    password="rootvale123",
    database="costos_variables",
)

TABLE_TCO = "costos_variables.tco"
TABLE_GEN = "costos_variables.prg_energia_horaria"
TABLE_MAP = "mapeo.homologa"

MAP_COL_TCO = "clave_tco"
MAP_COL_PO  = "clave_po"

GEN_EPS = 0.0
GEN_HOUR_COLS = [f"Hora{i}" for i in range(1, 25)]

SUBTIPOS_PONDERADOS = {"COGENERACION", "CARBÓN", "GNA", "GNL","BIOMASA"}

def read_tco_cons_esp(conn, fecha_inicio, fecha_fin) -> pd.DataFrame:
    sql = f"""
      SELECT
        {qident("Fecha")} AS Fecha,
        TRIM({qident("Central")}) AS Central_TCO,
        MAX({qident("Eficiencia")}) AS cons_esp,
        MAX(TRIM({qident("Subtipo")})) AS Subtipo
      FROM {TABLE_TCO}
      WHERE {qident("Fecha")} BETWEEN %s AND %s
        AND {qident("Central")} IS NOT NULL
        AND TRIM({qident("Central")}) <> ''
      GROUP BY {qident("Fecha")}, TRIM({qident("Central")})
    """
    df = pd.read_sql(sql, conn, params=[fecha_inicio, fecha_fin])
    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.date
    df["Central_TCO"] = df["Central_TCO"].astype(str).map(norm)
    df["cons_esp"] = pd.to_numeric(df["cons_esp"], errors="coerce")
    df["Subtipo"] = df["Subtipo"].astype(str).map(norm)
    return df


def cons_esp_pond_minmax_por_subtipo_horario(fecha_inicio, fecha_fin):
    with mysql.connector.connect(**MYSQL) as conn:
        mp = read_mapping_homologa(
            conn=conn,
            table_map=TABLE_MAP,
            col_src=MAP_COL_TCO,
            col_dst=MAP_COL_PO,
            src_alias="Central_TCO",
            dst_alias="Central_PO",
        )

        tco = read_tco_cons_esp(conn, fecha_inicio, fecha_fin)

        gen = read_hourly_long(
            conn=conn,
            table_name=TABLE_GEN,
            fecha_inicio=fecha_inicio,
            fecha_fin=fecha_fin,
            fecha_col="Fecha",
            entity_col="Central",
            hour_cols=GEN_HOUR_COLS,
            entity_alias="Central_PO",
            value_alias="gen",
        )

    # mapear TCO -> PO
    tco_m = tco.merge(mp, on="Central_TCO", how="left")

    tco_sin_mapeo = sorted(
        tco_m.loc[tco_m["Central_PO"].isna(), "Central_TCO"].drop_duplicates().tolist()
    )

    # consolidar TCO a nivel (Fecha, Central_PO) manteniendo Subtipo desde TCO
    ce_po = (
        tco_m.dropna(subset=["Central_PO"])
             .groupby(["Fecha", "Central_PO"], as_index=False)
             .agg(
                 cons_esp_po=("cons_esp", "mean"),
                 Subtipo=("Subtipo", "first"),
                 n_tco=("Central_TCO", "nunique"),
             )
    )

    # join con generación PRG
    base = gen.merge(ce_po, on=["Fecha", "Central_PO"], how="left")

    # ==================================================
    # PARTE A: SUBTIPOS PONDERADOS
    # - PON ponderado por gen
    # - MIN/MAX solo con gen > 0
    # ==================================================
    calc_pond = base.dropna(subset=["gen", "cons_esp_po", "Subtipo"]).copy()
    calc_pond = calc_pond.loc[calc_pond["gen"] > GEN_EPS].copy()
    calc_pond = calc_pond.loc[calc_pond["Subtipo"].isin(SUBTIPOS_PONDERADOS)].copy()

    if not calc_pond.empty:
        out_pond = weighted_min_max_by_group(
            calc_pond,
            group_cols=["Fecha", "hora", "Subtipo"],
            value_col="cons_esp_po",
            weight_col="gen",
            out_prefix="CE",
        )
        gen_pond = (
            calc_pond.groupby(["Fecha", "hora", "Subtipo"], as_index=False)
                     .agg(
                         GEN_MWh=("gen", "sum"),
                         n_centrales=("Central_PO", "nunique"),
                     )
        )
        out_pond = out_pond.merge(gen_pond, on=["Fecha", "hora", "Subtipo"], how="left")
    else:
        out_pond = pd.DataFrame(
            columns=["Fecha", "hora", "Subtipo", "CE_PON", "CE_MIN", "CE_MAX", "GEN_MWh", "n_centrales"]
        )

    # ==================================================
    # PARTE B: SUBTIPOS SIMPLES
    # - PON = promedio simple por subtipo desde TCO
    # - MIN/MAX simples desde TCO
    # - no dependen de generación
    # ==================================================
    tco_simple = tco.dropna(subset=["cons_esp", "Subtipo"]).copy()
    tco_simple = tco_simple.loc[~tco_simple["Subtipo"].isin(SUBTIPOS_PONDERADOS)].copy()

    if not tco_simple.empty:
        tco_simple_daily = (
            tco_simple.groupby(["Fecha", "Subtipo"], as_index=False)
                      .agg(
                          CE_PON=("cons_esp", "mean"),
                          CE_MIN=("cons_esp", "min"),
                          CE_MAX=("cons_esp", "max"),
                          n_centrales=("Central_TCO", "nunique"),
                      )
        )
    else:
        tco_simple_daily = pd.DataFrame(
            columns=["Fecha", "Subtipo", "CE_PON", "CE_MIN", "CE_MAX", "n_centrales"]
        )

    # generación por subtipo-hora solo para informar GEN_MWh
    base_simple_gen = base.dropna(subset=["gen", "Subtipo"]).copy()
    base_simple_gen = base_simple_gen.loc[base_simple_gen["gen"] > GEN_EPS].copy()
    base_simple_gen = base_simple_gen.loc[~base_simple_gen["Subtipo"].isin(SUBTIPOS_PONDERADOS)].copy()

    if not base_simple_gen.empty:
        gen_simple_h = (
            base_simple_gen.groupby(["Fecha", "hora", "Subtipo"], as_index=False)
                           .agg(GEN_MWh=("gen", "sum"))
        )
    else:
        gen_simple_h = pd.DataFrame(columns=["Fecha", "hora", "Subtipo", "GEN_MWh"])

    # expandir promedios diarios simples a las horas existentes
    horas = gen[["Fecha", "hora"]].drop_duplicates().copy()

    if not tco_simple_daily.empty and not horas.empty:
        horas["_k"] = 1
        tco_simple_daily["_k"] = 1
        out_simple = horas.merge(tco_simple_daily, on=["Fecha", "_k"], how="inner").drop(columns="_k")
        out_simple = out_simple.merge(gen_simple_h, on=["Fecha", "hora", "Subtipo"], how="left")
        out_simple["GEN_MWh"] = out_simple["GEN_MWh"].fillna(0.0)
    else:
        out_simple = pd.DataFrame(
            columns=["Fecha", "hora", "Subtipo", "CE_PON", "CE_MIN", "CE_MAX", "n_centrales", "GEN_MWh"]
        )

    # unir ambas partes
    out_long = pd.concat([out_pond, out_simple], ignore_index=True)
    out_long = out_long.sort_values(["Fecha", "hora", "Subtipo"]).reset_index(drop=True)

    if out_long.empty:
        out_wide = pd.DataFrame(columns=["Fecha", "hora"])
        diag = dict(
            total_tco=tco["Central_TCO"].nunique(),
            total_centrales_prg=gen["Central_PO"].nunique(),
            tco_sin_mapeo=len(tco_sin_mapeo),
            lista_tco_sin_mapeo=tco_sin_mapeo,
        )
        return out_wide, out_long, base, diag

    # pivot wide
    out_wide = pivot_three_metrics(
        out_long,
        index_cols=["Fecha", "hora"],
        subtipo_col="Subtipo",
        prefix="CE",
    )

    diag = dict(
        total_tco=tco["Central_TCO"].nunique(),
        total_centrales_prg=gen["Central_PO"].nunique(),
        tco_sin_mapeo=len(tco_sin_mapeo),
        lista_tco_sin_mapeo=tco_sin_mapeo,
    )

    return out_wide, out_long, base, diag


# RUN
ce_wide_df, ce_long_df, ce_base_df, diag = cons_esp_pond_minmax_por_subtipo_horario(
    FECHA_INICIO,
    FECHA_FIN
)

print("Diagnóstico:")
print(f"- Centrales únicas en TCO (rango): {diag['total_tco']}")
print(f"- Centrales únicas en PRG (rango): {diag['total_centrales_prg']}")
print(f"- TCO sin mapeo a PO: {diag['tco_sin_mapeo']}")

TABLE_OUT_SQL = "costos_variables.cons_esp_pond_minmax_subtipo_horario"
save_wide_to_mysql(ce_wide_df, TABLE_OUT_SQL, batch_size=500)
```

```python
import pandas as pd
import mysql.connector

TABLE_SRC = "costos_variables.tco"
TABLE_MAP = "mapeo.homologa"

VAL_COL = "CV"              # columna en tco
SRC_CENTRAL_COL = "Central" # columna en tco (nombre TCO)
SRC_FECHA_COL   = "Fecha"   # columna en tco

# Reusa tus variables existentes:
# MYSQL = {...}
# FECHA_INICIO, FECHA_FIN, N_HORAS, CENTRALES

def build_cv_from_tco_pandas(
    fecha_inicio,
    fecha_fin,
    centrales_prg_po: list[str],
    n_horas: int = 24,
    incluir_hora25: bool = False,
) -> pd.DataFrame:
    """
    Construye DF ancho (Fecha, hora, <clave_prg_po...>) usando:
      - tco: (Fecha, Central[TCO], CV)
      - homologa: mapea clave_tco -> clave_prg_po
    En salida, las columnas se llaman clave_prg_po (CENTRALES).
    Replica el valor diario a todas las horas.
    """
    horas = list(range(n_horas)) + ([24] if incluir_hora25 else [])
    horas_df = pd.DataFrame({"hora": horas}, dtype="int16")

    with mysql.connector.connect(**MYSQL) as conn:
        # 1) mapeo prg_po -> tco
        ph = ", ".join(["%s"] * len(centrales_prg_po))
        map_sql = f"""
            SELECT clave_po, clave_tco
            FROM {TABLE_MAP}
            WHERE clave_po IN ({ph})
              AND clave_tco IS NOT NULL AND clave_tco <> ''
        """
        hom = pd.read_sql(map_sql, conn, params=list(centrales_prg_po))
        if hom.empty:
            raise ValueError("No se encontró mapeo en homologa para las CENTRALES dadas (clave_po).")
        hom = hom.drop_duplicates(subset=["clave_po", "clave_tco"])

        # 2) traer TCO solo para claves_tco necesarias + rango fechas
        tco_keys = hom["clave_tco"].dropna().astype(str).unique().tolist()
        ph2 = ", ".join(["%s"] * len(tco_keys))
        src_sql = f"""
            SELECT {SRC_FECHA_COL} AS Fecha, {SRC_CENTRAL_COL} AS Central, {VAL_COL} AS valor
            FROM {TABLE_SRC}
            WHERE {SRC_FECHA_COL} BETWEEN %s AND %s
              AND {SRC_CENTRAL_COL} IN ({ph2})
        """
        src = pd.read_sql(src_sql, conn, params=[fecha_inicio, fecha_fin, *tco_keys])

    # 3) join para obtener clave_prg_po
    src["Fecha"] = pd.to_datetime(src["Fecha"]).dt.date
    src["Central"] = src["Central"].astype(str)

    df = src.merge(
        hom.rename(columns={"clave_po": "central_prg_po", "clave_tco": "central_tco"}),
        how="inner",
        left_on="Central",
        right_on="central_tco",
    )

    df = df[["Fecha", "central_prg_po", "valor"]]

    # 4) cross join con horas
    df["__k"] = 1
    horas_df["__k"] = 1
    dfh = df.merge(horas_df, on="__k", how="inner").drop(columns="__k")

    # 5) pivot ancho
    wide = (
        dfh.pivot_table(
            index=["Fecha", "hora"],
            columns="central_prg_po",
            values="valor",
            aggfunc="max",
        )
        .reset_index()
    )

    # 6) asegurar columnas finales y orden según centrales_prg_po
    for c in centrales_prg_po:
        if c not in wide.columns:
            wide[c] = pd.NA
    wide = wide[["Fecha", "hora"] + list(centrales_prg_po)]

    return wide

cv_df = build_cv_from_tco_pandas(
    fecha_inicio=FECHA_INICIO,
    fecha_fin=FECHA_FIN,
    centrales_prg_po=CENTRALES_TER,
    n_horas=N_HORAS,
    incluir_hora25=False,
)

# Guardado (elige una)
cv_df.to_pickle("cv.pkl")
# cv_df.to_parquet("cv.parquet", index=False)  # recomendado si tienes pyarrow
# cv_df.to_excel("cv.xlsx", index=False)       # ojo tamaño
```

```python
import pandas as pd
import mysql.connector
from datetime import date

MYSQL = dict(
    host="127.0.0.1",
    port=3306,
    user="root",
    password="rootvale123",
    database="costos_variables"
)

TABLE_PO   = "costos_variables.po"
TABLE_GEN  = "costos_variables.prg_energia_horaria"
TABLE_MAP  = "mapeo.homologa"

MAP_COL_PO   = "clave_po"
MAP_COL_SUB  = "Sub"

GEN_HOUR_COLS = [f"Hora{i}" for i in range(1, 25)]
GEN_EPS = 0.0

SUBTIPOS_PONDERADOS = {"COGENERACION", "CARBÓN", "CARBON", "GNA", "GNL","BIOMASA"}

def read_homologa_subtipo(conn) -> pd.DataFrame:
    sql = f"""
        SELECT
          TRIM({qident(MAP_COL_PO)}) AS Central,
          TRIM({qident(MAP_COL_SUB)}) AS Subtipo
        FROM {TABLE_MAP}
        WHERE {qident(MAP_COL_PO)} IS NOT NULL
          AND TRIM({qident(MAP_COL_PO)}) <> ''
          AND {qident(MAP_COL_SUB)} IS NOT NULL
          AND TRIM({qident(MAP_COL_SUB)}) <> ''
    """
    df = pd.read_sql(sql, conn)
    df["Central"] = df["Central"].astype(str).map(norm)
    df["Subtipo"] = df["Subtipo"].astype(str).map(norm)
    return df.drop_duplicates(subset=["Central", "Subtipo"])

def read_po_cvnc(conn, fecha_inicio, fecha_fin) -> pd.DataFrame:
    """
    PO diario por (Fecha, Central): solo CVNC.
    El subtipo se agregará desde homologa.
    """
    sql = f"""
        SELECT
          {qident("Fecha")} AS Fecha,
          TRIM({qident("Central")}) AS Central,
          MAX({qident("CVNC")}) AS cvnc
        FROM {TABLE_PO}
        WHERE {qident("Fecha")} BETWEEN %s AND %s
          AND {qident("Central")} IS NOT NULL
          AND TRIM({qident("Central")}) <> ''
        GROUP BY {qident("Fecha")}, TRIM({qident("Central")})
    """
    df = pd.read_sql(sql, conn, params=[fecha_inicio, fecha_fin])
    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.date
    df["Central"] = df["Central"].astype(str).map(norm)
    df["cvnc"] = pd.to_numeric(df["cvnc"], errors="coerce")
    return df


def cvnc_pond_minmax_por_subtipo_horario(fecha_inicio, fecha_fin):
    with mysql.connector.connect(**MYSQL) as conn:
        # 1) PO diario CVNC
        po = read_po_cvnc(conn, fecha_inicio, fecha_fin)

        # 2) Subtipo desde homologa
        homo = read_homologa_subtipo(conn)

        # 3) generación PRG horaria long
        gen = read_hourly_long(
            conn=conn,
            table_name=TABLE_GEN,
            fecha_inicio=fecha_inicio,
            fecha_fin=fecha_fin,
            fecha_col="Fecha",
            entity_col="Central",
            hour_cols=GEN_HOUR_COLS,
            entity_alias="Central",
            value_alias="gen",
        )

    # agregar subtipo desde homologa
    po = po.merge(homo, on="Central", how="left")

    # join directo por Fecha y Central
    base = gen.merge(po, on=["Fecha", "Central"], how="left")

    # diagnósticos
    centrales_gen_sin_po = sorted(
        base.loc[base["cvnc"].isna(), "Central"].dropna().drop_duplicates().tolist()
    )

    centrales_sin_sub_homologa = sorted(
        base.loc[base["Subtipo"].isna(), "Central"].dropna().drop_duplicates().tolist()
    )

    # ==================================================
    # PARTE A: SUBTIPOS PONDERADOS
    # ==================================================
    calc_pond = base.dropna(subset=["gen", "cvnc", "Subtipo"]).copy()
    calc_pond = calc_pond.loc[calc_pond["gen"] > GEN_EPS].copy()
    calc_pond = calc_pond.loc[calc_pond["Subtipo"].isin(SUBTIPOS_PONDERADOS)].copy()

    if not calc_pond.empty:
        out_pond = weighted_min_max_by_group(
            calc_pond,
            group_cols=["Fecha", "hora", "Subtipo"],
            value_col="cvnc",
            weight_col="gen",
            out_prefix="CVNC",
        )
        gen_pond = (
            calc_pond.groupby(["Fecha", "hora", "Subtipo"], as_index=False)
            .agg(GEN_MWh=("gen", "sum"), n_centrales=("Central", "nunique"))
        )
        out_pond = out_pond.merge(gen_pond, on=["Fecha", "hora", "Subtipo"], how="left")
    else:
        out_pond = pd.DataFrame(
            columns=["Fecha", "hora", "Subtipo", "CVNC_PON", "CVNC_MIN", "CVNC_MAX", "GEN_MWh", "n_centrales"]
        )

    # ==================================================
    # PARTE B: SUBTIPOS SIMPLES
    # ==================================================
    po_simple = po.dropna(subset=["cvnc", "Subtipo"]).copy()
    po_simple = po_simple.loc[~po_simple["Subtipo"].isin(SUBTIPOS_PONDERADOS)].copy()

    if not po_simple.empty:
        po_simple_daily = (
            po_simple.groupby(["Fecha", "Subtipo"], as_index=False)
            .agg(
                CVNC_PON=("cvnc", "mean"),
                CVNC_MIN=("cvnc", "min"),
                CVNC_MAX=("cvnc", "max"),
                n_centrales=("Central", "nunique"),
            )
        )
    else:
        po_simple_daily = pd.DataFrame(
            columns=["Fecha", "Subtipo", "CVNC_PON", "CVNC_MIN", "CVNC_MAX", "n_centrales"]
        )

    # generación total por subtipo-hora solo para informar GEN_MWh
    base_simple_gen = base.dropna(subset=["gen", "Subtipo"]).copy()
    base_simple_gen = base_simple_gen.loc[base_simple_gen["gen"] > GEN_EPS].copy()
    base_simple_gen = base_simple_gen.loc[~base_simple_gen["Subtipo"].isin(SUBTIPOS_PONDERADOS)].copy()

    if not base_simple_gen.empty:
        gen_simple_h = (
            base_simple_gen.groupby(["Fecha", "hora", "Subtipo"], as_index=False)
            .agg(GEN_MWh=("gen", "sum"))
        )
    else:
        gen_simple_h = pd.DataFrame(columns=["Fecha", "hora", "Subtipo", "GEN_MWh"])

    horas = gen[["Fecha", "hora"]].drop_duplicates().copy()

    if not po_simple_daily.empty and not horas.empty:
        horas["_k"] = 1
        po_simple_daily["_k"] = 1
        out_simple = horas.merge(po_simple_daily, on=["Fecha", "_k"], how="inner").drop(columns="_k")
        out_simple = out_simple.merge(gen_simple_h, on=["Fecha", "hora", "Subtipo"], how="left")
        out_simple["GEN_MWh"] = out_simple["GEN_MWh"].fillna(0.0)
    else:
        out_simple = pd.DataFrame(
            columns=["Fecha", "hora", "Subtipo", "CVNC_PON", "CVNC_MIN", "CVNC_MAX", "n_centrales", "GEN_MWh"]
        )

    # ==================================================
    # UNIÓN FINAL
    # ==================================================
    out_long = pd.concat([out_pond, out_simple], ignore_index=True)
    out_long = out_long.sort_values(["Fecha", "hora", "Subtipo"]).reset_index(drop=True)

    if out_long.empty:
        out_wide = pd.DataFrame(columns=["Fecha", "hora"])
        diag = dict(
            total_centrales_po=po["Central"].nunique(),
            total_centrales_gen=gen["Central"].nunique(),
            centrales_gen_sin_po=len(centrales_gen_sin_po),
            lista_gen_sin_po=centrales_gen_sin_po,
            centrales_sin_sub_homologa=len(centrales_sin_sub_homologa),
            lista_sin_sub_homologa=centrales_sin_sub_homologa,
        )
        return out_wide, out_long, base, diag

    out_wide = pivot_three_metrics(
        out_long,
        index_cols=["Fecha", "hora"],
        subtipo_col="Subtipo",
        prefix="CVNC",
    )

    diag = dict(
        total_centrales_po=po["Central"].nunique(),
        total_centrales_gen=gen["Central"].nunique(),
        centrales_gen_sin_po=len(centrales_gen_sin_po),
        lista_gen_sin_po=centrales_gen_sin_po,
        centrales_sin_sub_homologa=len(centrales_sin_sub_homologa),
        lista_sin_sub_homologa=centrales_sin_sub_homologa,
    )
    return out_wide, out_long, base, diag


# RUN
cvnc_wide_df, cvnc_long_df, cvnc_base_df, diag = cvnc_pond_minmax_por_subtipo_horario(
    FECHA_INICIO,
    FECHA_FIN
)

print("Diagnóstico CVNC:")
print(f"- Centrales únicas en PO (rango): {diag['total_centrales_po']}")
print(f"- Centrales únicas en PRG (rango): {diag['total_centrales_gen']}")
print(f"- Centrales PRG sin CVNC en PO: {diag['centrales_gen_sin_po']}")
print(f"- Centrales sin subtipo en homologa: {diag['centrales_sin_sub_homologa']}")

TABLE_OUT_SQL = "costos_variables.cvnc_pond_minmax_subtipo_horario"
save_wide_to_mysql(cvnc_wide_df, TABLE_OUT_SQL, batch_size=500)
```

```python
from __future__ import annotations

from pathlib import Path
from datetime import date
from typing import Dict, Tuple, List, Optional

import mysql.connector
from openpyxl import load_workbook


# =========================
# Config MySQL destino
# =========================
MYSQL = dict(
    host="127.0.0.1",
    port=3306,
    user="root",
    password="rootvale123",
)

DB_DEST = "planificacion"
TABLE_DEST = "reserva_conf"  # planificacion.reserva_conf


# =========================
# Util: resolver PRG con sufijos
# =========================
def _resolver_archivo_con_sufijos(path: str | Path) -> Path:
    path = Path(path)
    if path.exists():
        return path

    parent = path.parent
    stem = path.stem
    ext = path.suffix

    candidatos = [
        parent / f"{stem}{ext}",
        parent / f"{stem}-1{ext}",
        parent / f"{stem}-2{ext}",
    ]
    existentes = [p for p in candidatos if p.exists()]
    if existentes:
        return max(existentes, key=lambda p: p.stat().st_mtime)

    globbed = list(parent.glob(f"{stem}*{ext}"))
    if globbed:
        return max(globbed, key=lambda p: p.stat().st_mtime)

    raise FileNotFoundError(f"No se encontró: {path.name} (ni variantes -1/-2) en {parent}")


def _norm(x) -> str:
    if x is None:
        return ""
    s = str(x).replace("\u00A0", " ")
    s = " ".join(s.split())
    return s.strip()


# =========================
# 1) Fecha desde nombre PRG (YYMMDD -> 20YY-MM-DD)
# =========================
def fecha_desde_nombre_prg(prg_path: str | Path) -> date:
    """
    Extrae YYMMDD desde el nombre (ej: PRG250101.xlsx -> 2025-01-01).
    Toma los últimos 6 dígitos consecutivos del stem.
    """
    p = Path(prg_path)
    stem = p.stem  # PRG250101 o PRG250101-1
    digits = "".join(ch for ch in stem if ch.isdigit())
    if len(digits) < 6:
        raise ValueError(f"No pude extraer YYMMDD desde el nombre: {p.name}")
    yymmdd = digits[-6:]
    yy = int(yymmdd[0:2])
    mm = int(yymmdd[2:4])
    dd = int(yymmdd[4:6])
    return date(2000 + yy, mm, dd)


# =========================
# 2) Lectura rápida "* - Conf"
#    (evita ws.cell por celda: usa iter_rows con values_only)
# =========================
def _find_in_rows(
    rows: List[Tuple],
    needle: str,
    *,
    min_r: int,
    min_c: int,
    max_r: int,
    max_c: int
) -> Optional[Tuple[int, int]]:
    """
    Busca needle exacto (case-insensitive) en la subgrilla.
    rows es lista de tuplas 0-indexed.
    Retorna (r, c) 0-indexed.
    """
    needle_n = _norm(needle).lower()
    R = min(len(rows), max_r)
    for r in range(min_r, R):
        row = rows[r]
        C = min(len(row), max_c)
        for c in range(min_c, C):
            if _norm(row[c]).lower() == needle_n:
                return r, c
    return None


def leer_conf_rapido(
    prg_path: str | Path,
    sheet_name: str,
    tipo_reserva: str,
    *,
    start_row: int = 6,   # fila excel (1-indexed) donde está "CONFIGURACIÓN"
    start_col: int = 7,   # col excel (1-indexed) G
    n_horas: int = 24
) -> List[Tuple[date, int, str, str, float, float]]:
    """
    Lee hoja "* - Conf" y devuelve:
      (Fecha, hora(0-23), Configuracion, TipoReserva, Subida, Bajada)
    La fecha se obtiene del nombre del archivo (YYMMDD).
    """

    prg_path = _resolver_archivo_con_sufijos(prg_path)
    fecha = fecha_desde_nombre_prg(prg_path)

    wb = load_workbook(prg_path, read_only=True, data_only=True)
    if sheet_name not in wb.sheetnames:
        raise ValueError(f"{prg_path.name} no tiene hoja '{sheet_name}'. Hojas: {wb.sheetnames}")
    ws = wb[sheet_name]

    # Leer solo un área razonable para velocidad.
    # Ajusta si alguna hoja viene más grande.
    max_row = min(ws.max_row, 600)
    max_col = min(ws.max_column, 220)

    rows = list(ws.iter_rows(min_row=1, max_row=max_row, min_col=1, max_col=max_col, values_only=True))

    # Convertir start_row/start_col a 0-index
    sr = max(start_row - 1, 0)
    sc = max(start_col - 1, 0)

    # Encontrar CONFIGURACIÓN cerca del header
    pos_cfg = _find_in_rows(rows, "CONFIGURACIÓN", min_r=sr, min_c=sc, max_r=sr + 30, max_c=sc + 30)
    if pos_cfg:
        header_r, cfg_c = pos_cfg
    else:
        header_r, cfg_c = sr, sc  # fallback: asumir (fila 6, col G)

    # Encontrar SUBIDA / BAJADA (buscamos en las primeras ~80 filas desde col G)
    pos_sub = _find_in_rows(rows, "SUBIDA", min_r=0, min_c=sc, max_r=min(80, len(rows)), max_c=max_col)
    pos_baj = _find_in_rows(rows, "BAJADA", min_r=0, min_c=sc, max_r=min(80, len(rows)), max_c=max_col)
    if not pos_sub or not pos_baj:
        raise ValueError(f"No encontré 'SUBIDA' y/o 'BAJADA' en '{sheet_name}' (desde col {start_col}).")

    sub_r, sub_c0 = pos_sub
    baj_r, baj_c0 = pos_baj

    # Fila de horas suele ser 1 debajo del título
    horas_row_sub = sub_r + 1
    horas_row_baj = baj_r + 1

    def build_hour_map(row_idx: int, col_start_idx: int) -> Dict[int, int]:
        m: Dict[int, int] = {}
        if row_idx >= len(rows):
            return {h: col_start_idx + (h - 1) for h in range(1, n_horas + 1)}
        row = rows[row_idx]
        for j in range(n_horas):
            c = col_start_idx + j
            if c >= len(row):
                continue
            v = row[c]
            try:
                h = int(str(v).strip())
            except Exception:
                continue
            if 1 <= h <= n_horas:
                m[h] = c
        if len(m) < n_horas:
            m = {h: col_start_idx + (h - 1) for h in range(1, n_horas + 1)}
        return m

    sub_hmap = build_hour_map(horas_row_sub, sub_c0)
    baj_hmap = build_hour_map(horas_row_baj, baj_c0)

    def to_float(x) -> float:
        if x is None or _norm(x) == "":
            return 0.0
        try:
            return float(x)
        except Exception:
            return 0.0

    out: List[Tuple[date, int, str, str, float, float]] = []

    data_start = header_r + 1  # datos bajo "CONFIGURACIÓN"
    empty_streak = 0

    for r in range(data_start, len(rows)):
        row = rows[r]
        cfg = _norm(row[cfg_c] if cfg_c < len(row) else None)

        if not cfg:
            empty_streak += 1
            if empty_streak >= 5:
                break
            continue
        empty_streak = 0

        for h in range(1, n_horas + 1):
            cs = sub_hmap[h]
            cb = baj_hmap[h]
            v_sub = row[cs] if cs < len(row) else None
            v_baj = row[cb] if cb < len(row) else None
            out.append((fecha, h - 1, cfg, tipo_reserva, to_float(v_sub), to_float(v_baj)))

    return out


# =========================
# 3) SQL: tabla planificacion.reserva_conf + upsert
# =========================
def ensure_table_reserva_conf(conn):
    cur = conn.cursor()
    cur.execute(f"CREATE DATABASE IF NOT EXISTS `{DB_DEST}`;")
    cur.execute(f"""
    CREATE TABLE IF NOT EXISTS `{DB_DEST}`.`{TABLE_DEST}` (
      Fecha        DATE NOT NULL,
      hora         TINYINT NOT NULL,         -- 0..23
      Central VARCHAR(255) NOT NULL,
      TipoReserva  VARCHAR(20) NOT NULL,     -- Primaria/Secundaria/Terciaria
      SUBIDA       DECIMAL(18,6) NULL,
      BAJADA       DECIMAL(18,6) NULL,
      PRIMARY KEY (Fecha, hora, Configuracion, TipoReserva)
    ) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4;
    """)
    conn.commit()


def upsert_reserva_conf(prg_path: str | Path):
    prg_path = _resolver_archivo_con_sufijos(prg_path)

    # hojas -> tipo
    targets = [
        ("Reservas CPF - Conf", "Primaria", 6, 7),
        ("Reservas CSF - Conf", "Secundaria",6, 2),
        ("Reservas CTF - Conf", "Terciaria",6, 2),
    ]

    all_rows: List[Tuple[date, int, str, str, float, float]] = []
    for sheet, tipo, sr, sc in targets:
        all_rows.extend(
            leer_conf_rapido(
                prg_path,
                sheet,
                tipo,
                start_row=sr,
                start_col=sc,
                n_horas=24
            )
        )
    if not all_rows:
        print("Sin filas para insertar.")
        return

    conn = mysql.connector.connect(**MYSQL)
    try:
        ensure_table_reserva_conf(conn)
        cur = conn.cursor()

        sql = f"""
        INSERT INTO `{DB_DEST}`.`{TABLE_DEST}`
          (Fecha, hora, Central, TipoReserva, SUBIDA, BAJADA)
        VALUES (%s, %s, %s, %s, %s, %s)
        ON DUPLICATE KEY UPDATE
          SUBIDA = VALUES(SUBIDA),
          BAJADA = VALUES(BAJADA);
        """
        cur.executemany(sql, all_rows)
        conn.commit()

        print(f"OK: upsert {len(all_rows)} filas en {DB_DEST}.{TABLE_DEST} ({Path(prg_path).name})")
    finally:
        conn.close()


# =========================
# Uso
# =========================
upsert_reserva_conf("Programas/PRG/PRG250101.xlsx")
```

```python
from datetime import date, timedelta
from pathlib import Path

def _daterange(d1: date, d2: date):
    d = d1
    while d <= d2:
        yield d
        d += timedelta(days=1)

def upsert_reserva_conf_rango(
    fecha_inicio: date,
    fecha_fin: date,
    carpeta_prg: str | Path = "Programas/PRG",
    patron: str = "PRG{yy}{mm}{dd}.xlsx",
):
    """
    Recorre fechas [fecha_inicio..fecha_fin], busca PRG por fecha (con -1/-2) y sube a SQL.
    - carpeta_prg: carpeta donde están los PRG
    - patron: cómo se llama el archivo base (sin -1/-2). Por defecto PRGYYMMDD.xlsx
    """
    carpeta_prg = Path(carpeta_prg)

    ok = 0
    no_file = 0
    errors = 0

    for f in _daterange(fecha_inicio, fecha_fin):
        base = patron.format(yy=f.strftime("%y"), mm=f.strftime("%m"), dd=f.strftime("%d"))
        path = carpeta_prg / base
        try:
            upsert_reserva_conf(path)
            ok += 1
        except FileNotFoundError:
            print(f"[NO FILE] {f} -> {path.name}")
            no_file += 1
        except Exception as e:
            print(f"[ERROR] {f} -> {path.name}: {e}")
            errors += 1

    print(f"Resumen: OK={ok}, NO_FILE={no_file}, ERROR={errors}")


# -----------------------
# USO
# -----------------------
upsert_reserva_conf_rango(
     fecha_inicio=date(2026, 3, 3),
     fecha_fin=date(2026, 3, 28),
     carpeta_prg="Programas/PRG",
     patron="PRG{yy}{mm}{dd}.xlsx"
 )
```

```python
import re
import pandas as pd
import mysql.connector

# =========================
# CONFIG
# =========================
MYSQL = dict(host="127.0.0.1", port=3306, user="root", password="rootvale123")

T_SRC = "planificacion.reserva_conf"
DB_OUT = "planificacion"
T_OUT = "reservas_tec"

TIPO2PREF = {
    "Primaria": "CPF",
    "Secundaria": "CSF",
    "Terciaria": "CTF",
}

KEY_COLS = ["Fecha", "hora"]

# =========================
# HELPERS
# =========================
def q_ident(s: str) -> str:
    return "`" + s.replace("`", "") + "`"

def col_safe(s: str) -> str:
    s = (s or "").strip().replace("\u00A0", " ")
    s = s.translate(str.maketrans("áéíóúÁÉÍÓÚñÑ", "aeiouAEIOUnN"))
    s = re.sub(r"\s+", "_", s)
    s = re.sub(r"[^0-9A-Za-z_+\-]", "", s)
    if not s:
        s = "SIN_SUBTIPO"
    if s[0].isdigit():
        s = "_" + s
    return s

def subtipo_canonico(s: str) -> str:
    """
    Unifica nombres equivalentes de Subtipo a una versión canónica.
    """
    if s is None:
        return "SIN_SUBTIPO"

    s = str(s).strip().replace("\u00A0", " ")
    if s == "":
        return "SIN_SUBTIPO"

    # versión normalizada para comparar
    sn = s.translate(str.maketrans("áéíóúÁÉÍÓÚñÑ", "aeiouAEIOUnN"))
    sn = re.sub(r"\s+", " ", sn).strip().lower()

    mapa = {
        "carbon": "Carbón",
        "carbón": "Carbón",

        "diesel": "Diésel",
        "diésel": "Diésel",

        "bess": "BESS",

        "gas natural": "Gas Natural",
        "gna": "GNA",
        "gn": "GNA",
        "gnl": "GNL",
        "gnl inf": "GNL INF",
        "glp": "GLP",

        "solar": "Solar",
        "eolica": "Eólica",
        "eólica": "Eólica",
        "hidraulica": "Hidráulica",
        "hidráulica": "Hidráulica",
        "biomasa": "Biomasa",
        "biogas": "Biogás",
        "bio gas": "Biogás",
        "biogás": "Biogás",
        "petcoke": "PetCoke",
        "fuil oil": "Fuel Oil",
        "fuel oil": "Fuel Oil",
        "geotermica": "Geotérmica",
        "geotérmica": "Geotérmica",
        "propano": "Propano",
        "cs": "CS",
    }

    if sn in mapa:
        return mapa[sn]

    # fallback: dejar limpio y legible
    return " ".join(s.split())

def build_reservas_tec_df(conf: pd.DataFrame) -> pd.DataFrame:
    """
    Devuelve DF ancho con columnas:
      Fecha, hora,
      CPF+_<Subtipo>, CPF-_<Subtipo>, CSF+_<Subtipo>, CSF-_<Subtipo>, CTF+_<Subtipo>, CTF-_<Subtipo>
    donde '+' usa SUBIDA y '-' usa BAJADA.
    """
    df = conf.copy()

    df["TipoReserva"] = df["TipoReserva"].astype(str).str.strip()
    df["Subtipo"] = df["Subtipo"].apply(subtipo_canonico)

    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.date
    df["hora"] = pd.to_numeric(df["hora"], errors="coerce").fillna(0).astype(int)
    df["SUBIDA"] = pd.to_numeric(df["SUBIDA"], errors="coerce").fillna(0.0)
    df["BAJADA"] = pd.to_numeric(df["BAJADA"], errors="coerce").fillna(0.0)

    df = df[df["TipoReserva"].isin(TIPO2PREF.keys())].copy()
    if df.empty:
        return pd.DataFrame(columns=[*KEY_COLS])

    df["PREF"] = df["TipoReserva"].map(TIPO2PREF)

    # usar subtipo ya canónico
    g = (
        df.groupby(["Fecha", "hora", "PREF", "Subtipo"], as_index=False)[["SUBIDA", "BAJADA"]]
          .sum()
    )

    sub = g.pivot_table(
        index=["Fecha", "hora"],
        columns=["PREF", "Subtipo"],
        values="SUBIDA",
        aggfunc="sum",
        fill_value=0.0,
    )

    baj = g.pivot_table(
        index=["Fecha", "hora"],
        columns=["PREF", "Subtipo"],
        values="BAJADA",
        aggfunc="sum",
        fill_value=0.0,
    )

    def flatten_cols(multi_cols, sign: str) -> list[str]:
        out = []
        for pref, subt in multi_cols:
            out.append(f"{pref}{sign}_{col_safe(subt)}")
        return out

    sub.columns = flatten_cols(sub.columns, "+")
    baj.columns = flatten_cols(baj.columns, "-")

    wide = pd.concat([sub, baj], axis=1).reset_index()

    # seguridad extra: combinar columnas duplicadas si quedara alguna
    if wide.columns.duplicated().any():
        wide = wide.groupby(level=0, axis=1).sum()

    def sort_key(c: str):
        if c in ("Fecha", "hora"):
            return (0, c)
        m = re.match(r"^(CPF|CSF|CTF)([+\-])_", c)
        if not m:
            return (9, c)
        pref, sign = m.group(1), m.group(2)
        pref_order = {"CPF": 1, "CSF": 2, "CTF": 3}[pref]
        sign_order = {"+": 1, "-": 2}[sign]
        return (pref_order, sign_order, c)

    cols_sorted = sorted(wide.columns, key=sort_key)
    wide = wide[cols_sorted]

    return wide

def ensure_table(conn, df: pd.DataFrame):
    cur = conn.cursor()
    try:
        cur.execute(f"CREATE DATABASE IF NOT EXISTS `{DB_OUT}`;")

        value_cols = [c for c in df.columns if c not in KEY_COLS]
        cols_sql = ",\n  ".join([f"{q_ident(c)} DECIMAL(18,6) NULL" for c in value_cols])

        cur.execute(f"""
            CREATE TABLE IF NOT EXISTS `{DB_OUT}`.`{T_OUT}` (
              Fecha DATE NOT NULL,
              hora  TINYINT NOT NULL,
              {cols_sql},
              PRIMARY KEY (Fecha, hora)
            ) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4;
        """)

        for c in value_cols:
            try:
                cur.execute(f"ALTER TABLE `{DB_OUT}`.`{T_OUT}` ADD COLUMN {q_ident(c)} DECIMAL(18,6) NULL;")
            except mysql.connector.Error:
                pass

        conn.commit()
    finally:
        cur.close()

def upsert_df(conn, df: pd.DataFrame, chunk_size: int = 5000):
    if df.empty:
        print("DF vacío: nada que subir.")
        return

    key_cols = KEY_COLS
    value_cols = [c for c in df.columns if c not in key_cols]

    cols_insert = [*key_cols, *value_cols]
    placeholders = ", ".join(["%s"] * len(cols_insert))
    updates = ", ".join([f"{q_ident(c)}=VALUES({q_ident(c)})" for c in value_cols])

    sql = f"""
        INSERT INTO `{DB_OUT}`.`{T_OUT}` ({", ".join(q_ident(c) for c in cols_insert)})
        VALUES ({placeholders})
        ON DUPLICATE KEY UPDATE
          {updates};
    """

    cur = conn.cursor()
    try:
        rows = df.itertuples(index=False, name=None)
        batch = []
        n = 0

        for r in rows:
            r = list(r)
            for j, c in enumerate(df.columns):
                if c in key_cols:
                    continue
                v = r[j]
                r[j] = None if pd.isna(v) else float(v)

            batch.append(tuple(r))

            if len(batch) >= chunk_size:
                cur.executemany(sql, batch)
                conn.commit()
                n += len(batch)
                batch = []

        if batch:
            cur.executemany(sql, batch)
            conn.commit()
            n += len(batch)

        print(f"OK: upsert {n} filas en {DB_OUT}.{T_OUT}")
    finally:
        cur.close()

# =========================
# PIPELINE
# =========================
def run_build_and_upload():
    conn = mysql.connector.connect(**MYSQL)
    try:
        conf = pd.read_sql(
            f"SELECT Fecha, hora, TipoReserva, SUBIDA, BAJADA, Subtipo FROM {T_SRC}",
            conn
        )

        reservas_tec_df = build_reservas_tec_df(conf)

        # chequeo por si acaso
        dupes = reservas_tec_df.columns[reservas_tec_df.columns.duplicated()].tolist()
        if dupes:
            raise ValueError(f"Quedaron columnas duplicadas: {dupes}")

        ensure_table(conn, reservas_tec_df)
        upsert_df(conn, reservas_tec_df)

        print("Columnas creadas:", len(reservas_tec_df.columns))
        print("Primeras 15:", list(reservas_tec_df.columns[:15]))

    finally:
        conn.close()

run_build_and_upload()
```

```python
import pandas as pd
import numpy as np
import mysql.connector
from datetime import date

# =========================================================
# CONFIG
# =========================================================
MYSQL = dict(
    host="127.0.0.1",
    port=3306,
    user="root",
    password="rootvale123",
)

TABLE_TCO   = "costos_variables.tco"
TABLE_CMG   = "costos_variables.costos_marginales"
TABLE_GEN   = "costos_variables.prg_energia_horaria"
TABLE_MAP   = "mapeo.homologa"
TABLE_OUT   = "costos_variables.generacion_forzada_costo_por_tipo_horario"

MAP_COL_TCO   = "clave_tco"
MAP_COL_PRG   = "clave_po"
MAP_COL_BARRA = "Barra cmg"


GEN_HOUR_COLS = [f"Hora{i}" for i in range(1, 25)]     # prg_energia_horaria
TCO_HOUR_COLS = [f"Hora{i}" for i in range(1, 25)]     # tco
CMG_HOUR_COLS = [f"Hora{i}" for i in range(1, 25)]     # costos_marginales

TIPOS_ORDEN = [
    "GNA",
    "GNL",
    "CARBÓN",
    "DIÉSEL",
    "BIOGAS",
    "BIOMASA",
    "COGENERACIÓN",
    "CS",
    "FUIL OIL",
    "GEO",
    "GNL INF",
    "PETCOKE",
    "PROPANO",
]

# =========================================================
# UTILS
# =========================================================
def norm(s) -> str:
    s = str(s).replace("\u00A0", " ").strip().upper()
    s = " ".join(s.split())
    return s

def qident(col: str) -> str:
    return f"`{str(col).replace('`','``')}`"

def mysql_connect():
    return mysql.connector.connect(**MYSQL)

def classify_tipo(subtipo: str):
    s = norm(subtipo)

    mapping = {
        "GNA": "GNA",
        "GNL": "GNL",
        "CARBON": "CARBÓN",
        "CARBÓN": "CARBÓN",
        "DIESEL": "DIÉSEL",
        "DIÉSEL": "DIÉSEL",
        "BIOGAS": "BIOGAS",
        "BIOMASA": "BIOMASA",
        "COGENERACION": "COGENERACIÓN",
        "COGENERACIÓN": "COGENERACIÓN",
        "CS": "CS",
        "FUIL OIL": "FUIL OIL",
        "FUEL OIL": "FUIL OIL",
        "GEO": "GEO",
        "GEOTERMICA": "GEO",
        "GEOTÉRMICA": "GEO",
        "GNL INF": "GNL INF",
        "PETCOKE": "PETCOKE",
        "PROPANO": "PROPANO",
    }

    return mapping.get(s)

# =========================================================
# READS
# =========================================================
def read_mapping_homologa_tco_to_prg(conn) -> pd.DataFrame:
    sql = f"""
        SELECT
          TRIM({qident(MAP_COL_TCO)}) AS Central_TCO,
          TRIM({qident(MAP_COL_PRG)}) AS Central_PRG,
          TRIM({qident(MAP_COL_BARRA)}) AS Barra_Homologada
        FROM {TABLE_MAP}
        WHERE {qident(MAP_COL_TCO)} IS NOT NULL
          AND TRIM({qident(MAP_COL_TCO)}) <> ''
          AND {qident(MAP_COL_PRG)} IS NOT NULL
          AND TRIM({qident(MAP_COL_PRG)}) <> ''
    """
    df = pd.read_sql(sql, conn)

    df["Central_TCO"] = df["Central_TCO"].astype(str).map(norm)
    df["Central_PRG"] = df["Central_PRG"].astype(str).map(norm)
    df["Barra_Homologada"] = df["Barra_Homologada"].astype(str).map(norm)

    return df.drop_duplicates(subset=["Central_TCO", "Central_PRG", "Barra_Homologada"])

def read_tco_horaria_long(conn, fecha_inicio, fecha_fin) -> pd.DataFrame:
    parts = []
    for i, hc in enumerate(TCO_HOUR_COLS, start=1):
        hora_out = i - 1
        parts.append(f"""
        SELECT
          {qident("Fecha")} AS Fecha,
          {hora_out} AS hora,
          TRIM({qident("Central")}) AS Central_TCO,
          TRIM({qident("Subtipo")}) AS Subtipo,
          TRIM({qident("Nombre_Barra")}) AS Barra_TCO,
          COALESCE({qident(hc)}, 0) AS CV
        FROM {TABLE_TCO}
        WHERE {qident("Fecha")} BETWEEN %s AND %s
          AND {qident("Central")} IS NOT NULL
          AND TRIM({qident("Central")}) <> ''
        """.strip())

    sql = f"""
    SELECT
      Fecha,
      hora,
      Central_TCO,
      Subtipo,
      Barra_TCO,
      CV
    FROM (
      {" UNION ALL ".join(parts)}
    ) x
    """

    params = []
    for _ in TCO_HOUR_COLS:
        params.extend([fecha_inicio, fecha_fin])

    df = pd.read_sql(sql, conn, params=params)
    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.date
    df["hora"] = pd.to_numeric(df["hora"], errors="coerce").astype("Int64")
    df["Central_TCO"] = df["Central_TCO"].astype(str).map(norm)
    df["Subtipo"] = df["Subtipo"].astype(str).map(norm)
    df["Barra_TCO"] = df["Barra_TCO"].astype(str).map(norm)
    df["CV"] = pd.to_numeric(df["CV"], errors="coerce")

    return df

def read_prg_horaria_long(conn, fecha_inicio, fecha_fin) -> pd.DataFrame:
    parts = []
    for i, hc in enumerate(GEN_HOUR_COLS, start=1):
        hora_out = i - 1
        parts.append(f"""
        SELECT
          {qident("Fecha")} AS Fecha,
          {hora_out} AS hora,
          TRIM({qident("Central")}) AS Central_PRG,
          COALESCE({qident(hc)}, 0) AS gen
        FROM {TABLE_GEN}
        WHERE {qident("Fecha")} BETWEEN %s AND %s
          AND {qident("Central")} IS NOT NULL
          AND TRIM({qident("Central")}) <> ''
        """.strip())

    sql = f"""
    SELECT
      Fecha,
      hora,
      Central_PRG,
      SUM(gen) AS gen
    FROM (
      {" UNION ALL ".join(parts)}
    ) x
    GROUP BY Fecha, hora, Central_PRG
    """

    params = []
    for _ in GEN_HOUR_COLS:
        params.extend([fecha_inicio, fecha_fin])

    df = pd.read_sql(sql, conn, params=params)
    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.date
    df["hora"] = pd.to_numeric(df["hora"], errors="coerce").astype("Int64")
    df["Central_PRG"] = df["Central_PRG"].astype(str).map(norm)
    df["gen"] = pd.to_numeric(df["gen"], errors="coerce")

    return df

def read_cmg_horaria_long(conn, fecha_inicio, fecha_fin) -> pd.DataFrame:
    parts = []
    for i, hc in enumerate(CMG_HOUR_COLS, start=1):
        hora_out = i - 1
        parts.append(f"""
        SELECT
          {qident("Fecha")} AS Fecha,
          {hora_out} AS hora,
          TRIM({qident("Barra")}) AS Barra,
          COALESCE({qident(hc)}, 0) AS cmg
        FROM {TABLE_CMG}
        WHERE {qident("Fecha")} BETWEEN %s AND %s
          AND {qident("Barra")} IS NOT NULL
          AND TRIM({qident("Barra")}) <> ''
        """.strip())

    sql = f"""
    SELECT
      Fecha,
      hora,
      Barra,
      AVG(cmg) AS cmg
    FROM (
      {" UNION ALL ".join(parts)}
    ) x
    GROUP BY Fecha, hora, Barra
    """

    params = []
    for _ in CMG_HOUR_COLS:
        params.extend([fecha_inicio, fecha_fin])

    df = pd.read_sql(sql, conn, params=params)
    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.date
    df["hora"] = pd.to_numeric(df["hora"], errors="coerce").astype("Int64")
    df["Barra"] = df["Barra"].astype(str).map(norm)
    df["cmg"] = pd.to_numeric(df["cmg"], errors="coerce")

    return df

# =========================================================
# SQL WRITE
# =========================================================
def split_table_name(table_fq: str):
    parts = table_fq.split(".")
    if len(parts) == 2:
        return parts[0], parts[1]
    return None, parts[0]

def mysql_type_for_column(col_name: str) -> str:
    if col_name == "Fecha":
        return "DATE NOT NULL"
    if col_name == "hora":
        return "TINYINT NOT NULL"
    return "DOUBLE NULL"

def create_or_update_table_from_df(conn, df: pd.DataFrame, table_fq: str):
    schema, table = split_table_name(table_fq)
    cur = conn.cursor()

    cur.execute(f"""
    CREATE TABLE IF NOT EXISTS {qident(schema)}.{qident(table)} (
        {qident("Fecha")} DATE NOT NULL,
        {qident("hora")} TINYINT NOT NULL,
        PRIMARY KEY ({qident("Fecha")}, {qident("hora")})
    ) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4 COLLATE=utf8mb4_0900_ai_ci
    """)

    cur.execute(f"SHOW COLUMNS FROM {qident(schema)}.{qident(table)}")
    existing_cols = {row[0] for row in cur.fetchall()}

    for c in df.columns:
        if c not in existing_cols:
            cur.execute(
                f"ALTER TABLE {qident(schema)}.{qident(table)} "
                f"ADD COLUMN {qident(c)} {mysql_type_for_column(c)}"
            )

    conn.commit()
    cur.close()

def truncate_table(conn, table_fq: str):
    schema, table = split_table_name(table_fq)
    cur = conn.cursor()
    cur.execute(f"TRUNCATE TABLE {qident(schema)}.{qident(table)}")
    conn.commit()
    cur.close()

def save_wide_to_mysql(df: pd.DataFrame, table_fq: str, batch_size: int = 1000):
    if df.empty:
        raise ValueError("El DataFrame está vacío.")

    df2 = df.copy()
    df2["Fecha"] = pd.to_datetime(df2["Fecha"], errors="coerce").dt.date
    df2["hora"] = pd.to_numeric(df2["hora"], errors="coerce").astype("Int64")

    for c in df2.columns:
        if c not in ["Fecha", "hora"]:
            df2[c] = pd.to_numeric(df2[c], errors="coerce")

    df2 = df2.replace([np.inf, -np.inf], np.nan)
    df2 = df2.astype(object)
    df2 = df2.where(pd.notna(df2), None)

    def py_scalar(x):
        if x is None:
            return None
        if isinstance(x, np.integer):
            return int(x)
        if isinstance(x, np.floating):
            return float(x)
        if isinstance(x, pd.Timestamp):
            return x.date()
        return x

    conn = mysql_connect()
    try:
        create_or_update_table_from_df(conn, df2, table_fq)
        truncate_table(conn, table_fq)

        schema, table = split_table_name(table_fq)
        cols = list(df2.columns)

        sql = f"""
        INSERT INTO {qident(schema)}.{qident(table)}
        ({", ".join(qident(c) for c in cols)})
        VALUES ({", ".join(["%s"] * len(cols))})
        """

        rows = [
            tuple(py_scalar(v) for v in row)
            for row in df2[cols].itertuples(index=False, name=None)
        ]

        cur = conn.cursor()
        for i in range(0, len(rows), batch_size):
            cur.executemany(sql, rows[i:i+batch_size])
            conn.commit()
        cur.close()
    finally:
        conn.close()

# =========================================================
# MAIN
# =========================================================
def generacion_forzada_y_costo_por_tipo_horario(fecha_inicio, fecha_fin):
    with mysql_connect() as conn:
        mp  = read_mapping_homologa_tco_to_prg(conn)
        tco = read_tco_horaria_long(conn, fecha_inicio, fecha_fin)
        gen = read_prg_horaria_long(conn, fecha_inicio, fecha_fin)
        cmg = read_cmg_horaria_long(conn, fecha_inicio, fecha_fin)

    # TCO + homologa
    tco_m = tco.merge(mp, on="Central_TCO", how="left")

    tco_sin_mapeo = sorted(
        tco_m.loc[tco_m["Central_PRG"].isna(), "Central_TCO"].drop_duplicates().tolist()
    )

    tco_sin_barra_homologada = sorted(
        tco_m.loc[tco_m["Barra_Homologada"].isna(), "Central_TCO"].drop_duplicates().tolist()
    )

    # Solo TCO mapeadas a PRG
    tco_m = tco_m.dropna(subset=["Central_PRG"]).copy()

    # Tipo
    tco_m["TipoGen"] = tco_m["Subtipo"].map(classify_tipo)

    subtipos_sin_clasificar = sorted(
        tco_m.loc[tco_m["TipoGen"].isna(), "Subtipo"].drop_duplicates().tolist()
    )

    # PRG + TCO horario
    base = gen.merge(
        tco_m[
            [
                "Fecha",
                "hora",
                "Central_PRG",
                "Central_TCO",
                "Subtipo",
                "TipoGen",
                "Barra_TCO",
                "Barra_Homologada",
                "CV",
            ]
        ],
        on=["Fecha", "hora", "Central_PRG"],
        how="left"
    )

    # CMg por barra homologada
    base = base.merge(
        cmg.rename(columns={"Barra": "Barra_Homologada"}),
        on=["Fecha", "hora", "Barra_Homologada"],
        how="left"
    )

    # Generación forzada
    base["gen_forzada"] = np.where(
        (base["gen"].notna()) &
        (base["CV"].notna()) &
        (base["cmg"].notna()) &
        (base["gen"] > 0) &
        (base["CV"] > base["cmg"]),
        base["gen"],
        0.0
    )

    # Costo forzado
    base["costo_forzado"] = np.where(
        base["gen_forzada"] > 0,
        base["gen_forzada"] * (base["CV"] - base["cmg"]),
        0.0
    )

    calc = base.dropna(subset=["TipoGen"]).copy()

    out_long = (
        calc.groupby(["Fecha", "hora", "TipoGen"], as_index=False)
            .agg(
                GEN_FORZADA=("gen_forzada", "sum"),
                COSTO_FORZADO=("costo_forzado", "sum"),
            )
    )

    horas_base = base[["Fecha", "hora"]].drop_duplicates().copy()
    tipos_df = pd.DataFrame({"TipoGen": TIPOS_ORDEN})
    horas_base["_k"] = 1
    tipos_df["_k"] = 1
    grid = horas_base.merge(tipos_df, on="_k").drop(columns="_k")

    out_long = grid.merge(out_long, on=["Fecha", "hora", "TipoGen"], how="left")
    out_long["GEN_FORZADA"] = out_long["GEN_FORZADA"].fillna(0.0)
    out_long["COSTO_FORZADO"] = out_long["COSTO_FORZADO"].fillna(0.0)

    gen_w = (
        out_long.pivot_table(
            index=["Fecha", "hora"],
            columns="TipoGen",
            values="GEN_FORZADA",
            aggfunc="first"
        ).reset_index()
    )

    cost_w = (
        out_long.pivot_table(
            index=["Fecha", "hora"],
            columns="TipoGen",
            values="COSTO_FORZADO",
            aggfunc="first"
        ).reset_index()
    )

    gen_w = gen_w.rename(columns={
        c: f"{c}_GEN_FORZADA" for c in gen_w.columns if c not in ["Fecha", "hora"]
    })
    cost_w = cost_w.rename(columns={
        c: f"{c}_COSTO_FORZADO" for c in cost_w.columns if c not in ["Fecha", "hora"]
    })

    out_wide = (
        gen_w.merge(cost_w, on=["Fecha", "hora"], how="outer")
             .sort_values(["Fecha", "hora"])
             .reset_index(drop=True)
    )

    ordered = ["Fecha", "hora"]
    for t in TIPOS_ORDEN:
        ordered.append(f"{t}_GEN_FORZADA")
    for t in TIPOS_ORDEN:
        ordered.append(f"{t}_COSTO_FORZADO")

    for c in ordered:
        if c not in out_wide.columns:
            out_wide[c] = 0.0

    out_wide = out_wide[ordered]

    diag = dict(
        total_tco=tco["Central_TCO"].nunique(),
        total_centrales_prg=gen["Central_PRG"].nunique(),
        total_barras_cmg=cmg["Barra"].nunique(),
        tco_sin_mapeo=len(tco_sin_mapeo),
        lista_tco_sin_mapeo=tco_sin_mapeo,
        tco_sin_barra_homologada=len(tco_sin_barra_homologada),
        lista_tco_sin_barra_homologada=tco_sin_barra_homologada,
        subtipos_sin_clasificar=subtipos_sin_clasificar,
        filas_con_cmg=len(base.loc[base["cmg"].notna()]),
        filas_sin_cmg=len(base.loc[base["cmg"].isna()]),
    )

    return out_wide, out_long, base, diag

# =========================================================
# RUN
# =========================================================
out_wide_df, out_long_df, base_df, diag = generacion_forzada_y_costo_por_tipo_horario(
    FECHA_INICIO,
    FECHA_FIN
)

print("Diagnóstico:")
print(f"- Centrales únicas en TCO: {diag['total_tco']}")
print(f"- Centrales únicas en PRG: {diag['total_centrales_prg']}")
print(f"- Barras únicas en CMg: {diag['total_barras_cmg']}")
print(f"- TCO sin mapeo: {diag['tco_sin_mapeo']}")
print(f"- Subtipos sin clasificar: {diag['subtipos_sin_clasificar']}")

save_wide_to_mysql(out_wide_df, TABLE_OUT, batch_size=1000)
```

```python
import pandas as pd
import numpy as np
import mysql.connector

def split_table_name(table_fq: str):
    parts = table_fq.split(".")
    if len(parts) == 2:
        return parts[0], parts[1]
    return MYSQL["database"], parts[0]

def qident(col: str) -> str:
    return f"`{str(col).replace('`','``')}`"

def mysql_connect():
    return mysql.connector.connect(**MYSQL)

def mysql_type_from_series(col_name: str, s: pd.Series) -> str:
    if col_name == "Fecha":
        return "DATE NULL"
    if col_name == "hora":
        return "TINYINT NULL"

    if pd.api.types.is_integer_dtype(s):
        return "BIGINT NULL"
    if pd.api.types.is_float_dtype(s):
        return "DOUBLE NULL"
    if pd.api.types.is_bool_dtype(s):
        return "TINYINT(1) NULL"

    # texto / mixto
    max_len = s.dropna().astype(str).map(len).max() if not s.dropna().empty else 10
    if pd.isna(max_len):
        max_len = 10

    if max_len <= 100:
        return "VARCHAR(100) NULL"
    if max_len <= 255:
        return "VARCHAR(255) NULL"
    if max_len <= 1000:
        return "VARCHAR(1000) NULL"
    return "TEXT NULL"

def create_table_from_df_generic(conn, df: pd.DataFrame, table_fq: str):
    schema, table = split_table_name(table_fq)

    cols_sql = []
    for c in df.columns:
        col_type = mysql_type_from_series(c, df[c])
        cols_sql.append(f"{qident(c)} {col_type}")

    sql = f"""
    CREATE TABLE IF NOT EXISTS {qident(schema)}.{qident(table)} (
        {", ".join(cols_sql)}
    ) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4 COLLATE=utf8mb4_0900_ai_ci
    """

    cur = conn.cursor()
    cur.execute(sql)
    conn.commit()
    cur.close()

def truncate_table(conn, table_fq: str):
    schema, table = split_table_name(table_fq)
    cur = conn.cursor()
    cur.execute(f"TRUNCATE TABLE {qident(schema)}.{qident(table)}")
    conn.commit()
    cur.close()

def save_df_to_mysql_generic(df: pd.DataFrame, table_fq: str, batch_size: int = 1000):
    if df.empty:
        raise ValueError("El DataFrame está vacío.")

    df2 = df.copy()

    # normalizar tipos
    if "Fecha" in df2.columns:
        df2["Fecha"] = pd.to_datetime(df2["Fecha"], errors="coerce").dt.date
    if "hora" in df2.columns:
        df2["hora"] = pd.to_numeric(df2["hora"], errors="coerce").astype("Int64")

    # reemplazar inf por nan
    df2 = df2.replace([np.inf, -np.inf], np.nan)

    # pasar a object para permitir None reales
    df2 = df2.astype(object)
    df2 = df2.where(pd.notna(df2), None)

    def py_scalar(x):
        if x is None:
            return None
        if isinstance(x, np.integer):
            return int(x)
        if isinstance(x, np.floating):
            return float(x)
        if isinstance(x, pd.Timestamp):
            return x.to_pydatetime()
        return x

    conn = mysql_connect()
    try:
        create_table_from_df_generic(conn, df2, table_fq)
        truncate_table(conn, table_fq)

        schema, table = split_table_name(table_fq)
        cols = list(df2.columns)

        sql = f"""
        INSERT INTO {qident(schema)}.{qident(table)}
        ({", ".join(qident(c) for c in cols)})
        VALUES ({", ".join(["%s"] * len(cols))})
        """

        rows = [
            tuple(py_scalar(v) for v in row)
            for row in df2[cols].itertuples(index=False, name=None)
        ]

        cur = conn.cursor()
        for i in range(0, len(rows), batch_size):
            cur.executemany(sql, rows[i:i+batch_size])
            conn.commit()
        cur.close()

    finally:
        conn.close()
```

```python
import pandas as pd
import mysql.connector
from functools import reduce
from openpyxl import load_workbook
import numpy as np

def build_master_time_keys(blocks: list[pd.DataFrame], key_cols: list[str]) -> pd.DataFrame:
    keys_base = None

    for b in blocks:
        if b.empty:
            continue

        cols_presentes = [c for c in key_cols if c in b.columns]
        if len(cols_presentes) != len(key_cols):
            continue

        k = b[key_cols].drop_duplicates().copy()

        if keys_base is None:
            keys_base = k
        else:
            keys_base = pd.concat([keys_base, k], ignore_index=True)

    if keys_base is None:
        return pd.DataFrame(columns=key_cols)

    keys_base = keys_base.drop_duplicates().sort_values(key_cols).reset_index(drop=True)
    return keys_base

def align_block_to_master_keys(df: pd.DataFrame, master_keys: pd.DataFrame, key_cols: list[str]) -> pd.DataFrame:
    if df.empty:
        return master_keys.copy()

    cols_presentes = [c for c in key_cols if c in df.columns]
    if len(cols_presentes) != len(key_cols):
        return master_keys.copy()

    out = master_keys.merge(df, on=key_cols, how="left")
    out = out.loc[:, ~out.columns.duplicated()]
    return out

def drop_key_cols(df: pd.DataFrame, key_cols: list[str]) -> pd.DataFrame:
    return df.drop(columns=[c for c in key_cols if c in df.columns], errors="ignore")

def reorder_forzada_cols(df: pd.DataFrame, subtipos: list[str], key_cols: list[str]) -> pd.DataFrame:
    """
    Fuerza orden:
      (keys) + [Subtipo_GEN_FORZADA...] + [Subtipo_COSTO_FORZADO...]
    Si una columna no existe, la crea en blanco.
    Luego deja el resto al final.
    """
    if df.empty:
        cols = [c for c in key_cols]
        for s in subtipos:
            cols.append(f"{s}_GEN_FORZADA")
        for s in subtipos:
            cols.append(f"{s}_COSTO_FORZADO")
        return pd.DataFrame(columns=cols)

    out = df.copy()
    keys = [c for c in key_cols if c in out.columns]

    desired = []

    for s in subtipos:
        col = f"{s}_GEN_FORZADA"
        desired.append(col)
        if col not in out.columns:
            out[col] = pd.NA

    for s in subtipos:
        col = f"{s}_COSTO_FORZADO"
        desired.append(col)
        if col not in out.columns:
            out[col] = pd.NA

    rest = [c for c in out.columns if c not in keys and c not in desired]
    out = out[keys + desired + rest]
    out = out.loc[:, ~out.columns.duplicated()]
    return out
def add_date_parts(df: pd.DataFrame, fecha_col: str = "Fecha", hora_col: str = "hora") -> pd.DataFrame:
    df = df.copy()
    if fecha_col not in df.columns:
        return df

    f = pd.to_datetime(df[fecha_col], errors="coerce")
    df[fecha_col] = f.dt.date
    df["Año"] = f.dt.year
    df["Mes"] = f.dt.month
    df["Día"] = f.dt.day

    cols = list(df.columns)

    ordered = []
    if fecha_col in cols:
        ordered.append(fecha_col)
    for c in ["Año", "Mes", "Día"]:
        if c in cols:
            ordered.append(c)
    if hora_col in cols:
        ordered.append(hora_col)

    rest = [c for c in cols if c not in ordered]
    return df[ordered + rest]

def normalize_numeric_series(s: pd.Series) -> pd.Series:
    """
    Convierte a numérico manejando:
    - números ya numéricos
    - strings con coma decimal
    - strings con punto decimal
    - strings con separador de miles
    - strings con espacios
    """
    if pd.api.types.is_numeric_dtype(s):
        return pd.to_numeric(s, errors="coerce")

    x = s.astype(str).str.strip()

    # vacíos / null-like
    x = x.replace({
        "": pd.NA,
        "nan": pd.NA,
        "None": pd.NA,
        "NaN": pd.NA,
        "<NA>": pd.NA,
        "nat": pd.NA,
        "NaT": pd.NA,
    })

    def _fix_num(v):
        if pd.isna(v):
            return pd.NA
        v = str(v).strip()

        # si tiene coma y punto, decidir decimal por la última aparición
        if "," in v and "." in v:
            if v.rfind(",") > v.rfind("."):
                # 1.234,56 -> 1234.56
                v = v.replace(".", "").replace(",", ".")
            else:
                # 1,234.56 -> 1234.56
                v = v.replace(",", "")
        elif "," in v:
            # 123,45 -> 123.45
            v = v.replace(",", ".")
        # si solo tiene punto, se deja tal cual

        return v

    x = x.map(_fix_num)
    return pd.to_numeric(x, errors="coerce")

def coerce_non_key_columns_to_numeric(df: pd.DataFrame, key_cols: list[str], blank_prefix: str = "__blank_") -> pd.DataFrame:
    df = df.copy()
    for c in df.columns:
        if c in key_cols:
            continue
        if str(c).startswith(blank_prefix):
            continue
        df[c] = normalize_numeric_series(df[c]).round(1)
    return df

def mysql_conn(schema: str | None = None):
    return mysql.connector.connect(
        host=MYSQL_HOST, port=MYSQL_PORT,
        user=MYSQL_USER, password=MYSQL_PASSWORD,
        database=schema
    )

def _build_where_date_ranges(fecha_col: str, date_ranges):
    parts = []
    params = []
    for ini, fin in date_ranges:
        parts.append(f"({fecha_col} >= %s AND {fecha_col} < %s)")
        params.extend([ini, fin])
    return " OR ".join(parts), params

def read_sql_table(table_fq: str) -> pd.DataFrame:
    sql = f"SELECT * FROM {table_fq}"
    conn = mysql_conn()
    try:
        return pd.read_sql(sql, conn)
    finally:
        conn.close()

def read_sql_table_filtered(table_fq: str, date_ranges, fecha_col="Fecha") -> pd.DataFrame:
    where_sql, params = _build_where_date_ranges(fecha_col, date_ranges)
    sql = f"SELECT * FROM {table_fq} WHERE {where_sql}"
    conn = mysql_conn()
    try:
        return pd.read_sql(sql, conn, params=params)
    finally:
        conn.close()

def dedup_preserve_order(lst):
    seen = set()
    out = []
    for x in lst:
        if x not in seen:
            seen.add(x)
            out.append(x)
    return out

def apply_column_order(df: pd.DataFrame, desired: list[str]) -> pd.DataFrame:
    if not desired:
        return df
    desired_existing = [c for c in desired if c in df.columns]
    rest = [c for c in df.columns if c not in desired_existing]
    return df[desired_existing + rest]

def apply_time_block_order(df: pd.DataFrame, desired: list[str], key_cols: list[str]) -> pd.DataFrame:
    if df.empty:
        return df
    keys = [c for c in key_cols if c in df.columns]
    desired = dedup_preserve_order(desired)
    desired_existing = [c for c in desired if c in df.columns and c not in keys]
    rest = [c for c in df.columns if c not in keys and c not in desired_existing]
    out = df[keys + desired_existing + rest]
    out = out.loc[:, ~out.columns.duplicated()]
    return out

def filter_time_ranges(df: pd.DataFrame, date_ranges, fecha_col="Fecha") -> pd.DataFrame:
    if fecha_col not in df.columns:
        return df
    out = df.copy()
    out[fecha_col] = pd.to_datetime(out[fecha_col], errors="coerce")

    masks = []
    for ini, fin in date_ranges:
        ini_dt = pd.to_datetime(ini)
        fin_dt = pd.to_datetime(fin)
        masks.append((out[fecha_col] >= ini_dt) & (out[fecha_col] < fin_dt))

    if not masks:
        return out.iloc[0:0]

    mask = masks[0]
    for m in masks[1:]:
        mask = mask | m

    out = out.loc[mask].copy()
    out[fecha_col] = out[fecha_col].dt.date
    return out

def prefix_non_key_columns(df: pd.DataFrame, prefix: str, key_cols: list[str]) -> pd.DataFrame:
    df = df.copy()
    ren = {c: f"{prefix}{c}" for c in df.columns if c not in key_cols}
    return df.rename(columns=ren)

def outer_merge_on_keys(dfs: list[pd.DataFrame], key_cols: list[str]) -> pd.DataFrame:
    def _merge(a, b):
        return a.merge(b, on=key_cols, how="outer")
    return reduce(_merge, dfs)

_blank_counter = 0
def blank_cols(n_rows: int, n: int = 1) -> pd.DataFrame:
    global _blank_counter
    _blank_counter += 1
    pref = f"{_blank_counter}_"
    return pd.DataFrame({f"__blank_{pref}{i+1}__": [pd.NA] * n_rows for i in range(n)})

def side_by_side_blocks(blocks: list[pd.DataFrame], blank_between: int = 0) -> pd.DataFrame:
    fixed_blocks = []
    for b in blocks:
        bb = b.copy()
        if bb.columns.duplicated().any():
            bb = bb.loc[:, ~bb.columns.duplicated()]
        fixed_blocks.append(bb)

    max_rows = max((len(b) for b in fixed_blocks), default=0)
    padded = []
    for b in fixed_blocks:
        bb = b.copy()
        if len(bb) < max_rows:
            pad = pd.DataFrame({c: [pd.NA] * (max_rows - len(bb)) for c in bb.columns})
            bb = pd.concat([bb, pad], ignore_index=True)
        padded.append(bb)

    out_parts = []
    for i, b in enumerate(padded):
        out_parts.append(b)
        if i < len(padded) - 1:
            out_parts.append(blank_cols(max_rows, blank_between))
    out = pd.concat(out_parts, axis=1)
    out = out.loc[:, ~out.columns.duplicated()]
    return out

def build_pond_order(subtipos: list[str]) -> list[str]:
    return [f"{s}_{m}" for s in subtipos for m in ("PON", "MIN", "MAX")]

def reorder_pond_cols(df: pd.DataFrame, subtipos: list[str], key_cols: list[str]) -> pd.DataFrame:
    """
    Fuerza orden:
      (keys) + [Subtipo_PON, Subtipo_MIN, Subtipo_MAX] para cada subtipo en 'subtipos'
    Si una columna no existe, la crea en blanco.
    Luego deja el resto al final.
    """
    if df.empty:
        # incluso si está vacío, crear estructura esperada
        cols = [c for c in key_cols]
        for s in subtipos:
            for suf in ("PON", "MIN", "MAX"):
                cols.append(f"{s}_{suf}")
        return pd.DataFrame(columns=cols)

    out = df.copy()
    keys = [c for c in key_cols if c in out.columns]

    desired = []
    for s in subtipos:
        for suf in ("PON", "MIN", "MAX"):
            col = f"{s}_{suf}"
            desired.append(col)
            if col not in out.columns:
                out[col] = pd.NA

    rest = [c for c in out.columns if c not in keys and c not in desired]
    out = out[keys + desired + rest]
    out = out.loc[:, ~out.columns.duplicated()]
    return out

def write_df_into_existing_excel(df: pd.DataFrame,
                                 path: str,
                                 sheet_name: str,
                                 start_row: int = 5,
                                 start_col: int = 1,
                                 clear_old_area: bool = False):
    wb = load_workbook(path)
    if sheet_name not in wb.sheetnames:
        ws = wb.create_sheet(sheet_name)
    else:
        ws = wb[sheet_name]

    # ✅ convertir pd.NA/NaN/Inf a None (openpyxl lo soporta)
    df2 = df.copy()

    # todo a object para permitir None
    df2 = df2.astype(object)

    # reemplazos robustos
    df2 = df2.where(pd.notna(df2), None)                  # NaN, NaT, pd.NA -> None
    df2 = df2.replace([np.inf, -np.inf], None)

    nrows = len(df2) + 1  # + header
    ncols = len(df2.columns)

    if clear_old_area and nrows > 0 and ncols > 0:
        for r in range(start_row, start_row + nrows):
            for c in range(start_col, start_col + ncols):
                ws.cell(row=r, column=c).value = None

    # header
    for j, colname in enumerate(df2.columns, start=start_col):
        ws.cell(row=start_row, column=j).value = str(colname)

    # data
    values = df2.to_numpy(dtype=object)
    for i in range(values.shape[0]):
        for j in range(values.shape[1]):
            ws.cell(row=start_row + 1 + i, column=start_col + j).value = values[i, j]

    wb.save(path)

def align_blocks_on_keys_without_repeating_keys(blocks: list[pd.DataFrame], key_cols: list[str], blank_between: int = 1) -> pd.DataFrame:
    if not blocks:
        return pd.DataFrame()

    keys_base = None
    for b in blocks:
        k = b[key_cols].drop_duplicates()
        if keys_base is None:
            keys_base = k.copy()
        else:
            keys_base = keys_base.merge(k, on=key_cols, how="outer")

    keys_base = keys_base.sort_values(key_cols).reset_index(drop=True)

    aligned_blocks = [keys_base]
    for b in blocks:
        bb = keys_base.merge(b, on=key_cols, how="left")
        bb = bb.drop(columns=[c for c in key_cols if c in bb.columns], errors="ignore")
        bb = bb.loc[:, ~bb.columns.duplicated()]
        aligned_blocks.append(bb)

    return side_by_side_blocks(aligned_blocks, blank_between=blank_between)
```

```python
import pandas as pd
import mysql.connector
from openpyxl import load_workbook
import numpy as np
import os

# -----------------------
# Config MySQL
# -----------------------
MYSQL_HOST = "127.0.0.1"
MYSQL_PORT = 3306
MYSQL_USER = "root"
MYSQL_PASSWORD = "rootvale123"

date_ranges = [
    ("2025-01-01", "2025-03-29"),
    ("2026-01-01", "2026-03-29"),
]

# -----------------------
# Tablas SQL
# -----------------------
T_SUBTIPO_OR       = "datos_operacion_real.subtipo"
T_CENTRALES_OR     = "datos_operacion_real.centrales"
T_BARRAS_OR        = "datos_operacion_real.barras"
T_SUBTIPOS_PLAN    = "planificacion.subtipos"
T_BARRAS_PLAN_HORA = "planificacion.barras"
T_RESERVAS_TEC     = "planificacion.reservas_tec"

T_COMB_POND        = "costos_variables.combustible_pond_minmax_subtipo_horario"
T_CE_POND          = "costos_variables.cons_esp_pond_minmax_subtipo_horario"
T_CV_POND          = "costos_variables.cv_pond_minmax_subtipo_horario"
T_CVNC_POND        = "costos_variables.cvnc_pond_minmax_subtipo_horario"
T_GEN_FORZADA      = "costos_variables.generacion_forzada_costo_por_tipo_horario"
T_COMB_EMPRESA     = "costos_variables.combustible_pond_minmax_empresa"

# -----------------------
# Pickles
# -----------------------
P_CENTRALES_PRG = "centrales_prg_energia_horaria.pkl"
P_PRECIO_PO     = "precio_combustible_po.pkl"
P_CONS_ESP      = "cons_esp.pkl"
P_CV            = "cv.pkl"
P_CVNC          = "cvnc.pkl"

# -----------------------
# Orden columnas
# -----------------------
ORDER_SUBTIPO_OR = ["Solar","Termosolar","Eólica","Geotérmica","Pasada","Embalse","BioGas","Biomasa",
                    "Cogeneracion","PetCoke","Carbón","Gas Natural","Diésel","Fuel Oil","Térmica",
                    "Inyección","Retiro","GLP"]

ORDER_CENTRALES_OR = [
    "Nehuenco Gas Natural","Nehuenco II Gas Natural","Nehuenco 9B Gas Natural","Candelaria 1 Gas Arg","Candelaria 2 Gas Arg",
    "Nueva Renca Gas Arg","San Isidro Gas Arg","San Isidro II Gas Arg","Quintero TG 1A Gas Arg","Quintero TG 1B Gas Arg",
    "Los Vientos Gas Arg","TER Colmito Gas Argentino","Coronel TG Gas Natural","CMPC Cordillera Gas Arg","Newen Gas Arg",
    "TER Yungay U1 Gas Argentino","TER Yungay U2 Gas Argentino","TER Yungay U3 Gas Argentino","Atacama TG1A Gas Arg",
    "Atacama TG2A Gas Arg","Taltal 1 Gas Arg","Taltal 2 Gas Arg","KELAR TG1 (Gas Arg)","CTM3-TG Gas Arg",
    "TER TOCOPILLA TG3 Gas Arg","U16-TG Gas Arg","Nehuenco GNL","Nehuenco II GNL","Nehuenco 9B GNL","Candelaria 1 GNL",
    "Candelaria 2 GNL","Nueva Renca GNL","Nueva Renca FA GNL","San Isidro GNL","San Isidro II GNL","Quintero TG 1A GNL",
    "Quintero TG 1B GNL","Los Vientos GNL","TER Colmito GNL","Coronel TG GNL","CMPC Cordillera GNL","Newen GNL",
    "TER Yungay U1 GNL","TER Yungay U2 GNL","TER Yungay U3 GNL","HBS-GNL","Atacama TG1A GNL","Atacama TG2A GNL",
    "Taltal 1 GNL","Taltal 2 GNL","KELAR TG1 (GNL)","CTM3-TG GNL","TER TOCOPILLA TG3 GNL","U16-TG GNL",
    "Nueva Renca FA GLP","CMPC Tissue","PMGD TER BIO CRUZ","Lipigas Concón","PINARES","Tapihue","PMGD TER TOMAVAL",
    "Newen Propano","TER ANGAMOS-ANG1","TER ANGAMOS-ANG2","TER COCHRANE-CCR1 (Carbon)","TER COCHRANE-CCR2 (Carbon)", "TER ANDINA", "TER HORNITOS",
    "TER IEM","CTM1","CTM2","TER Guacolda U1","TER Guacolda U2","TER Guacolda U3","TER Guacolda U4","TER Guacolda U5","TER CAMPICHE",
    "TER NUEVA VENTANAS","Santa María","Atacama TG1A Diésel","Atacama TG2A Diésel","Taltal 1 Diésel","Taltal 2 Diésel",
    "CTM3-TG Diesel","TER TOCOPILLA TG1","TER TOCOPILLA TG2","TER TOCOPILLA TG3 Diesel","U16-TG Diesel","KELAR TG1 (Diesel)",
    "Candelaria 1 Diésel","Candelaria 2 Diésel","Los Vientos Diésel","Nehuenco Diésel","Nehuenco 9B Diésel","Nehuenco II Diésel",
    "Nueva Renca Diésel","Quintero TG 1A Diésel","Quintero TG 1B Diésel","San Isidro Diésel","San Isidro II Diésel",
    "TER Colmito Diésel","Santa Lidia","Otros","EL TORO","CANUTILLAR","CIPRESES","RAPEL","COLBÚN","PANGUE",
    "PEHUENCHE","RALCO","MACHICURA",
]

ORDER_BARRAS_OR     = ["A_JAHUEL","CRUCERO","MAITENCILLO","P_MONTT","QUILLOTA","CHARRUA","D_ALMAGRO"]
ORDER_BARRAS_PL     = ["A.JAHUEL","CRUCERO","MAITENCILLO","P.MONTT","QUILLOTA","CHARRUA","D.ALMAGRO"]
ORDER_SUBTIPOS_PLAN = ["Pasada","Eólica","Solar","Embalse","Diésel","Fuil Oil","Geotermica","GNA","GNL",
                       "Carbon","Cogeneracion","Biomasa","BioGas","GNL INF","GLP","PROPANO","PetCoke",
                       "Inyección","Retiro","Ver"]

ORDER_SUBTIPOS_POND = ["BIOGAS","BIOMASA","COGENERACION","CS","DIÉSEL","FUEL OIL","GEO","GLP","PETCOKE","PROPANO","BESS","CARBÓN","GNL","GNA"]
ORDER_SUBTIPOS_FORZ = ["BIOGAS","BIOMASA","CARBÓN","COGENERACIÓN","CS","DIÉSEL",
                       "FUIL OIL","GEO","GLP","GNA","GNL","GNL INF","PETCOKE","PROPANO"]

PREF_PRG        = "PRG_"
PREF_BPL        = "BPL_"
PREF_PC         = "PC_"
PREF_CE         = "CE_"
PREF_CV         = "CV_"
PREF_CVNC       = "CVNC_"
PREF_RTEC       = "RTEC_"
PREF_GF         = "GF_"
PREF_COMB_POND  = "COMB_"
PREF_CE_POND    = "CE_POND_"
PREF_CV_POND    = "CV_POND_"
PREF_CVNC_POND  = "CVNC_POND_"
PREF_COMB_EMP   = "COMB_EMP_"

MERGE_KEY_COLS   = ["Fecha", "hora"]
DISPLAY_KEY_COLS = ["Fecha", "Año", "Mes", "Día", "hora"]

EXCEL_PATH = "unificada_completa.xlsx"
SHEET_NAME = "unificada"
START_ROW  = 5
START_COL  = 1

# -----------------------
# Helpers
# -----------------------
def add_date_parts(df: pd.DataFrame, fecha_col: str = "Fecha", hora_col: str = "hora") -> pd.DataFrame:
    df = df.copy()
    if fecha_col not in df.columns:
        return df
    f = pd.to_datetime(df[fecha_col], errors="coerce")
    df[fecha_col] = f.dt.date
    df["Año"] = f.dt.year
    df["Mes"] = f.dt.month
    df["Día"] = f.dt.day
    cols = list(df.columns)
    ordered = []
    if fecha_col in cols:
        ordered.append(fecha_col)
    for c in ["Año", "Mes", "Día"]:
        if c in cols:
            ordered.append(c)
    if hora_col in cols:
        ordered.append(hora_col)
    rest = [c for c in cols if c not in ordered]
    return df[ordered + rest]

def normalize_numeric_series(s: pd.Series) -> pd.Series:
    if pd.api.types.is_numeric_dtype(s):
        return pd.to_numeric(s, errors="coerce")
    x = s.astype(str).str.strip()
    x = x.replace({
        "": pd.NA, "nan": pd.NA, "None": pd.NA, "NaN": pd.NA,
        "<NA>": pd.NA, "nat": pd.NA, "NaT": pd.NA,
    })
    def _fix_num(v):
        if pd.isna(v):
            return pd.NA
        v = str(v).strip()
        if "," in v and "." in v:
            if v.rfind(",") > v.rfind("."):
                v = v.replace(".", "").replace(",", ".")
            else:
                v = v.replace(",", "")
        elif "," in v:
            v = v.replace(",", ".")
        return v
    x = x.map(_fix_num)
    return pd.to_numeric(x, errors="coerce")

def coerce_non_key_columns_to_numeric(df: pd.DataFrame, key_cols: list[str], blank_prefix: str = "__blank_") -> pd.DataFrame:
    df = df.copy()
    for c in df.columns:
        if c in key_cols:
            continue
        if str(c).startswith(blank_prefix):
            continue
        df[c] = normalize_numeric_series(df[c]).round(1)
    return df

def mysql_conn(schema: str | None = None):
    return mysql.connector.connect(
        host=MYSQL_HOST, port=MYSQL_PORT,
        user=MYSQL_USER, password=MYSQL_PASSWORD,
        database=schema
    )

def _build_where_date_ranges(fecha_col: str, date_ranges):
    parts = []
    params = []
    for ini, fin in date_ranges:
        parts.append(f"({fecha_col} >= %s AND {fecha_col} < %s)")
        params.extend([ini, fin])
    return " OR ".join(parts), params

def read_sql_table_filtered(table_fq: str, date_ranges, fecha_col="Fecha") -> pd.DataFrame:
    where_sql, params = _build_where_date_ranges(fecha_col, date_ranges)
    sql = f"SELECT * FROM {table_fq} WHERE {where_sql}"
    conn = mysql_conn()
    try:
        return pd.read_sql(sql, conn, params=params)
    finally:
        conn.close()

def dedup_preserve_order(lst):
    seen = set()
    out = []
    for x in lst:
        if x not in seen:
            seen.add(x)
            out.append(x)
    return out

def apply_column_order(df: pd.DataFrame, desired: list[str]) -> pd.DataFrame:
    if not desired:
        return df
    desired_existing = [c for c in desired if c in df.columns]
    rest = [c for c in df.columns if c not in desired_existing]
    return df[desired_existing + rest]

def apply_time_block_order(df: pd.DataFrame, desired: list[str], key_cols: list[str]) -> pd.DataFrame:
    if df.empty:
        return df
    keys = [c for c in key_cols if c in df.columns]
    desired = dedup_preserve_order(desired)
    desired_existing = [c for c in desired if c in df.columns and c not in keys]
    rest = [c for c in df.columns if c not in keys and c not in desired_existing]
    out = df[keys + desired_existing + rest]
    return out.loc[:, ~out.columns.duplicated()]

def filter_time_ranges(df: pd.DataFrame, date_ranges, fecha_col="Fecha") -> pd.DataFrame:
    if fecha_col not in df.columns:
        return df
    out = df.copy()
    out[fecha_col] = pd.to_datetime(out[fecha_col], errors="coerce")
    masks = []
    for ini, fin in date_ranges:
        ini_dt = pd.to_datetime(ini)
        fin_dt = pd.to_datetime(fin)
        masks.append((out[fecha_col] >= ini_dt) & (out[fecha_col] < fin_dt))
    if not masks:
        return out.iloc[0:0]
    mask = masks[0]
    for m in masks[1:]:
        mask = mask | m
    out = out.loc[mask].copy()
    out[fecha_col] = out[fecha_col].dt.date
    return out

def prefix_non_key_columns(df: pd.DataFrame, prefix: str, key_cols: list[str]) -> pd.DataFrame:
    df = df.copy()
    ren = {c: f"{prefix}{c}" for c in df.columns if c not in key_cols}
    return df.rename(columns=ren)

_blank_counter = 0
def blank_cols(n_rows: int, n: int = 1) -> pd.DataFrame:
    global _blank_counter
    _blank_counter += 1
    pref = f"{_blank_counter}_"
    return pd.DataFrame({f"__blank_{pref}{i+1}__": [pd.NA] * n_rows for i in range(n)})

def side_by_side_blocks(blocks: list[pd.DataFrame], blank_between: int = 0) -> pd.DataFrame:
    fixed_blocks = []
    for b in blocks:
        bb = b.copy()
        if bb.columns.duplicated().any():
            bb = bb.loc[:, ~bb.columns.duplicated()]
        fixed_blocks.append(bb)

    max_rows = max((len(b) for b in fixed_blocks), default=0)
    padded = []
    for b in fixed_blocks:
        bb = b.copy()
        if len(bb) < max_rows:
            pad = pd.DataFrame({c: [pd.NA] * (max_rows - len(bb)) for c in bb.columns})
            bb = pd.concat([bb, pad], ignore_index=True)
        padded.append(bb)

    out_parts = []
    for i, b in enumerate(padded):
        out_parts.append(b)
        if i < len(padded) - 1 and blank_between > 0:
            out_parts.append(blank_cols(max_rows, blank_between))
    out = pd.concat(out_parts, axis=1)
    return out.loc[:, ~out.columns.duplicated()]

def build_pond_order(subtipos: list[str]) -> list[str]:
    return [f"{s}_{m}" for s in subtipos for m in ("PON", "MIN", "MAX")]

def reorder_pond_cols(df: pd.DataFrame, subtipos: list[str], key_cols: list[str]) -> pd.DataFrame:
    if df.empty:
        cols = [c for c in key_cols]
        for s in subtipos:
            for suf in ("PON", "MIN", "MAX"):
                cols.append(f"{s}_{suf}")
        return pd.DataFrame(columns=cols)

    out = df.copy()
    keys = [c for c in key_cols if c in out.columns]
    desired = []
    for s in subtipos:
        for suf in ("PON", "MIN", "MAX"):
            col = f"{s}_{suf}"
            desired.append(col)
            if col not in out.columns:
                out[col] = pd.NA
    rest = [c for c in out.columns if c not in keys and c not in desired]
    out = out[keys + desired + rest]
    return out.loc[:, ~out.columns.duplicated()]

def reorder_forzada_cols(df: pd.DataFrame, subtipos: list[str], key_cols: list[str]) -> pd.DataFrame:
    if df.empty:
        cols = [c for c in key_cols]
        for s in subtipos:
            cols.append(f"{s}_GEN_FORZADA")
        for s in subtipos:
            cols.append(f"{s}_COSTO_FORZADO")
        return pd.DataFrame(columns=cols)

    out = df.copy()
    keys = [c for c in key_cols if c in out.columns]
    desired = []
    for s in subtipos:
        col = f"{s}_GEN_FORZADA"
        desired.append(col)
        if col not in out.columns:
            out[col] = pd.NA
    for s in subtipos:
        col = f"{s}_COSTO_FORZADO"
        desired.append(col)
        if col not in out.columns:
            out[col] = pd.NA
    rest = [c for c in out.columns if c not in keys and c not in desired]
    out = out[keys + desired + rest]
    return out.loc[:, ~out.columns.duplicated()]

def build_master_time_keys(blocks: list[pd.DataFrame], key_cols: list[str]) -> pd.DataFrame:
    keys_base = None
    for b in blocks:
        if b.empty:
            continue
        cols_presentes = [c for c in key_cols if c in b.columns]
        if len(cols_presentes) != len(key_cols):
            continue
        k = b[key_cols].drop_duplicates().copy()
        if keys_base is None:
            keys_base = k
        else:
            keys_base = pd.concat([keys_base, k], ignore_index=True)
    if keys_base is None:
        return pd.DataFrame(columns=key_cols)
    return keys_base.drop_duplicates().sort_values(key_cols).reset_index(drop=True)

def align_block_to_master_keys(df: pd.DataFrame, master_keys: pd.DataFrame, key_cols: list[str]) -> pd.DataFrame:
    if df.empty:
        return master_keys.copy()
    cols_presentes = [c for c in key_cols if c in df.columns]
    if len(cols_presentes) != len(key_cols):
        return master_keys.copy()
    out = master_keys.merge(df, on=key_cols, how="left")
    return out.loc[:, ~out.columns.duplicated()]

def drop_key_cols(df: pd.DataFrame, key_cols: list[str]) -> pd.DataFrame:
    return df.drop(columns=[c for c in key_cols if c in df.columns], errors="ignore")

def assemble_blocks_flexible(block_specs: list[tuple[pd.DataFrame, int]]) -> pd.DataFrame:
    """
    block_specs = [(df_bloque, n_blank_after), ...]
    Inserta n_blank_after columnas vacías después de cada bloque.
    """
    fixed_blocks = []
    max_rows = max((len(df) for df, _ in block_specs), default=0)

    for df, _ in block_specs:
        bb = df.copy()
        if bb.columns.duplicated().any():
            bb = bb.loc[:, ~bb.columns.duplicated()]
        if len(bb) < max_rows:
            pad = pd.DataFrame({c: [pd.NA] * (max_rows - len(bb)) for c in bb.columns})
            bb = pd.concat([bb, pad], ignore_index=True)
        fixed_blocks.append(bb)

    parts = []
    for i, ((_, n_blank), bb) in enumerate(zip(block_specs, fixed_blocks)):
        parts.append(bb)
        if n_blank > 0:
            parts.append(blank_cols(max_rows, n_blank))

    out = pd.concat(parts, axis=1) if parts else pd.DataFrame()
    return out.loc[:, ~out.columns.duplicated()]

def write_df_to_new_excel(df: pd.DataFrame,
                          path: str,
                          sheet_name: str = "unificada",
                          start_row: int = 5,
                          start_col: int = 1):
    from openpyxl import Workbook

    wb = Workbook()
    ws = wb.active
    ws.title = sheet_name

    df2 = df.copy().astype(object)
    df2 = df2.where(pd.notna(df2), None)
    df2 = df2.replace([np.inf, -np.inf], None)

    print("Guardando archivo NUEVO en:", os.path.abspath(path))
    print("Shape:", df2.shape)

    for j, colname in enumerate(df2.columns, start=start_col):
        ws.cell(row=start_row, column=j).value = str(colname)

    values = df2.to_numpy(dtype=object)
    for i in range(values.shape[0]):
        for j in range(values.shape[1]):
            cell = ws.cell(row=start_row + 1 + i, column=start_col + j)
            val = values[i, j]
            colname = df2.columns[j]
            cell.value = val

            if colname == "Fecha":
                cell.number_format = "yyyy-mm-dd"
            elif colname in ("Año", "Mes", "Día", "hora"):
                cell.number_format = "0"
            elif str(colname).startswith("__blank_"):
                pass
            else:
                cell.number_format = "0.0"

    wb.save(path)
    wb.close()
    print("Archivo creado correctamente.")

# -----------------------
# 1) Cargar tablas SQL base
# -----------------------
subtipo_or_df    = read_sql_table_filtered(T_SUBTIPO_OR, date_ranges, fecha_col="Fecha")
centrales_or_df  = read_sql_table_filtered(T_CENTRALES_OR, date_ranges, fecha_col="Fecha")
barras_or_df     = read_sql_table_filtered(T_BARRAS_OR, date_ranges, fecha_col="Fecha")
subtipos_plan_df = read_sql_table_filtered(T_SUBTIPOS_PLAN, date_ranges, fecha_col="Fecha")

subtipo_or_df    = apply_time_block_order(subtipo_or_df, ORDER_SUBTIPO_OR, MERGE_KEY_COLS)
centrales_or_df  = apply_time_block_order(centrales_or_df, ORDER_CENTRALES_OR, MERGE_KEY_COLS)
barras_or_df     = apply_time_block_order(barras_or_df, ORDER_BARRAS_OR, MERGE_KEY_COLS)
subtipos_plan_df = apply_column_order(subtipos_plan_df, ORDER_SUBTIPOS_PLAN)
subtipos_plan_df = subtipos_plan_df.loc[:, ~subtipos_plan_df.columns.duplicated()]

# -----------------------
# 2) Pickles + tablas plan
# -----------------------
centrales_prg_df = pd.read_pickle(P_CENTRALES_PRG)
precio_po_df     = pd.read_pickle(P_PRECIO_PO)
cons_esp_df      = pd.read_pickle(P_CONS_ESP)
cv_df            = pd.read_pickle(P_CV)
cvnc_df          = pd.read_pickle(P_CVNC)
barras_plan_df   = read_sql_table_filtered(T_BARRAS_PLAN_HORA, date_ranges, fecha_col="Fecha")

centrales_prg_df = filter_time_ranges(centrales_prg_df, date_ranges, fecha_col="Fecha")
precio_po_df     = filter_time_ranges(precio_po_df, date_ranges, fecha_col="Fecha")
cons_esp_df      = filter_time_ranges(cons_esp_df, date_ranges, fecha_col="Fecha")
cv_df            = filter_time_ranges(cv_df, date_ranges, fecha_col="Fecha")
cvnc_df          = filter_time_ranges(cvnc_df, date_ranges, fecha_col="Fecha")

barras_plan_df   = apply_time_block_order(barras_plan_df, ORDER_BARRAS_PL, MERGE_KEY_COLS)

centrales_prg_df = prefix_non_key_columns(centrales_prg_df, PREF_PRG, MERGE_KEY_COLS)
subtipos_plan_df = prefix_non_key_columns(subtipos_plan_df, PREF_PRG, MERGE_KEY_COLS)
barras_plan_df   = prefix_non_key_columns(barras_plan_df, PREF_BPL, MERGE_KEY_COLS)
precio_po_df     = prefix_non_key_columns(precio_po_df, PREF_PC, MERGE_KEY_COLS)
cons_esp_df      = prefix_non_key_columns(cons_esp_df, PREF_CE, MERGE_KEY_COLS)
cv_df            = prefix_non_key_columns(cv_df, PREF_CV, MERGE_KEY_COLS)
cvnc_df          = prefix_non_key_columns(cvnc_df, PREF_CVNC, MERGE_KEY_COLS)

# -----------------------
# 3) Tablas wide SQL
# -----------------------
pond_order = build_pond_order(ORDER_SUBTIPOS_POND)

comb_pond_df    = read_sql_table_filtered(T_COMB_POND, date_ranges, fecha_col="Fecha")
ce_pond_df      = read_sql_table_filtered(T_CE_POND, date_ranges, fecha_col="Fecha")
cv_pond_df      = read_sql_table_filtered(T_CV_POND, date_ranges, fecha_col="Fecha")
cvnc_pond_df    = read_sql_table_filtered(T_CVNC_POND, date_ranges, fecha_col="Fecha")
gen_forzada_df  = read_sql_table_filtered(T_GEN_FORZADA, date_ranges, fecha_col="Fecha")
comb_empresa_df = read_sql_table_filtered(T_COMB_EMPRESA, date_ranges, fecha_col="Fecha")

comb_pond_df   = apply_time_block_order(comb_pond_df, pond_order, MERGE_KEY_COLS)
ce_pond_df     = apply_time_block_order(ce_pond_df, pond_order, MERGE_KEY_COLS)
cv_pond_df     = apply_time_block_order(cv_pond_df, pond_order, MERGE_KEY_COLS)
cvnc_pond_df   = apply_time_block_order(cvnc_pond_df, pond_order, MERGE_KEY_COLS)
gen_forzada_df = reorder_forzada_cols(gen_forzada_df, ORDER_SUBTIPOS_FORZ, MERGE_KEY_COLS)

comb_pond_df   = reorder_pond_cols(comb_pond_df, ORDER_SUBTIPOS_POND, MERGE_KEY_COLS)
ce_pond_df     = reorder_pond_cols(ce_pond_df, ORDER_SUBTIPOS_POND, MERGE_KEY_COLS)
cv_pond_df     = reorder_pond_cols(cv_pond_df, ORDER_SUBTIPOS_POND, MERGE_KEY_COLS)
cvnc_pond_df   = reorder_pond_cols(cvnc_pond_df, ORDER_SUBTIPOS_POND, MERGE_KEY_COLS)

comb_pond_df    = prefix_non_key_columns(comb_pond_df, PREF_COMB_POND, MERGE_KEY_COLS)
ce_pond_df      = prefix_non_key_columns(ce_pond_df, PREF_CE_POND, MERGE_KEY_COLS)
cv_pond_df      = prefix_non_key_columns(cv_pond_df, PREF_CV_POND, MERGE_KEY_COLS)
cvnc_pond_df    = prefix_non_key_columns(cvnc_pond_df, PREF_CVNC_POND, MERGE_KEY_COLS)
gen_forzada_df  = prefix_non_key_columns(gen_forzada_df, PREF_GF, MERGE_KEY_COLS)
comb_empresa_df = prefix_non_key_columns(comb_empresa_df, PREF_COMB_EMP, MERGE_KEY_COLS)

# -----------------------
# 4) Reservas
# -----------------------
reservas_tec_df = read_sql_table_filtered(T_RESERVAS_TEC, date_ranges, fecha_col="Fecha")
reservas_tec_df = apply_time_block_order(reservas_tec_df, desired=[], key_cols=MERGE_KEY_COLS)
reservas_tec_df = prefix_non_key_columns(reservas_tec_df, PREF_RTEC, MERGE_KEY_COLS)

if not reservas_tec_df.empty:
    for c in reservas_tec_df.columns:
        if c not in MERGE_KEY_COLS:
            reservas_tec_df[c] = pd.to_numeric(reservas_tec_df[c], errors="coerce")

    for g in RTEC_GROUPS:
        pref = f"{PREF_RTEC}{g}_"
        out_col = f"{PREF_RTEC}{g}_Otros"
        group_cols = [c for c in reservas_tec_df.columns if c.startswith(pref)]
        std_cols = [c for c in group_cols if c.split(pref, 1)[1] in (RTEC_KNOWN_SUFFIX - {"Otros"})]
        extra_cols = [c for c in group_cols if c not in std_cols and c != out_col]

        if out_col not in reservas_tec_df.columns:
            reservas_tec_df[out_col] = 0.0

        reservas_tec_df[out_col] = reservas_tec_df[extra_cols].fillna(0).sum(axis=1) if extra_cols else 0.0

for c in ORDER_RTEC:
    if c not in reservas_tec_df.columns:
        reservas_tec_df[c] = pd.NA

reservas_tec_df = reservas_tec_df[
    MERGE_KEY_COLS
    + [c for c in ORDER_RTEC if c in reservas_tec_df.columns]
    + [c for c in reservas_tec_df.columns if c not in MERGE_KEY_COLS and c not in ORDER_RTEC]
]
reservas_tec_df = reservas_tec_df.loc[:, ~reservas_tec_df.columns.duplicated()]

# -----------------------
# 5) Base maestra temporal
# -----------------------
all_time_blocks = [
    subtipo_or_df,
    centrales_or_df,
    barras_or_df,
    subtipos_plan_df,
    centrales_prg_df,
    barras_plan_df,
    precio_po_df,
    cons_esp_df,
    cv_df,
    cvnc_df,
    comb_pond_df,
    ce_pond_df,
    cv_pond_df,
    cvnc_pond_df,
    reservas_tec_df,
    gen_forzada_df,
    comb_empresa_df,
    
]

master_time_keys_df = build_master_time_keys(all_time_blocks, MERGE_KEY_COLS)

subtipo_or_df    = align_block_to_master_keys(subtipo_or_df, master_time_keys_df, MERGE_KEY_COLS)
centrales_or_df  = align_block_to_master_keys(centrales_or_df, master_time_keys_df, MERGE_KEY_COLS)
barras_or_df     = align_block_to_master_keys(barras_or_df, master_time_keys_df, MERGE_KEY_COLS)
subtipos_plan_df = align_block_to_master_keys(subtipos_plan_df, master_time_keys_df, MERGE_KEY_COLS)
centrales_prg_df = align_block_to_master_keys(centrales_prg_df, master_time_keys_df, MERGE_KEY_COLS)
barras_plan_df   = align_block_to_master_keys(barras_plan_df, master_time_keys_df, MERGE_KEY_COLS)
precio_po_df     = align_block_to_master_keys(precio_po_df, master_time_keys_df, MERGE_KEY_COLS)
cons_esp_df      = align_block_to_master_keys(cons_esp_df, master_time_keys_df, MERGE_KEY_COLS)
cv_df            = align_block_to_master_keys(cv_df, master_time_keys_df, MERGE_KEY_COLS)
cvnc_df          = align_block_to_master_keys(cvnc_df, master_time_keys_df, MERGE_KEY_COLS)
comb_pond_df     = align_block_to_master_keys(comb_pond_df, master_time_keys_df, MERGE_KEY_COLS)
ce_pond_df       = align_block_to_master_keys(ce_pond_df, master_time_keys_df, MERGE_KEY_COLS)
cv_pond_df       = align_block_to_master_keys(cv_pond_df, master_time_keys_df, MERGE_KEY_COLS)
cvnc_pond_df     = align_block_to_master_keys(cvnc_pond_df, master_time_keys_df, MERGE_KEY_COLS)
reservas_tec_df  = align_block_to_master_keys(reservas_tec_df, master_time_keys_df, MERGE_KEY_COLS)
comb_empresa_df  = align_block_to_master_keys(comb_empresa_df, master_time_keys_df, MERGE_KEY_COLS)
gen_forzada_df   = align_block_to_master_keys(gen_forzada_df, master_time_keys_df, MERGE_KEY_COLS)

# -----------------------
# 6) Convertir y armar TODO dentro de time_unificada_df
# -----------------------
subtipo_or_df     = add_date_parts(subtipo_or_df)
centrales_or_df   = add_date_parts(centrales_or_df)
barras_or_df      = add_date_parts(barras_or_df)
subtipos_plan_df  = add_date_parts(subtipos_plan_df)
centrales_prg_df  = add_date_parts(centrales_prg_df)
barras_plan_df    = add_date_parts(barras_plan_df)
precio_po_df      = add_date_parts(precio_po_df)
cons_esp_df       = add_date_parts(cons_esp_df)
cv_df             = add_date_parts(cv_df)
cvnc_df           = add_date_parts(cvnc_df)
comb_pond_df      = add_date_parts(comb_pond_df)
ce_pond_df        = add_date_parts(ce_pond_df)
cv_pond_df        = add_date_parts(cv_pond_df)
cvnc_pond_df      = add_date_parts(cvnc_pond_df)
reservas_tec_df   = add_date_parts(reservas_tec_df)
comb_empresa_df   = add_date_parts(comb_empresa_df)
gen_forzada_df    = add_date_parts(gen_forzada_df)

export_key_cols = ["Fecha", "Año", "Mes", "Día", "hora"]

subtipo_or_df     = coerce_non_key_columns_to_numeric(subtipo_or_df, export_key_cols)
centrales_or_df   = coerce_non_key_columns_to_numeric(centrales_or_df, export_key_cols)
barras_or_df      = coerce_non_key_columns_to_numeric(barras_or_df, export_key_cols)
subtipos_plan_df  = coerce_non_key_columns_to_numeric(subtipos_plan_df, export_key_cols)
centrales_prg_df  = coerce_non_key_columns_to_numeric(centrales_prg_df, export_key_cols)
barras_plan_df    = coerce_non_key_columns_to_numeric(barras_plan_df, export_key_cols)
precio_po_df      = coerce_non_key_columns_to_numeric(precio_po_df, export_key_cols)
cons_esp_df       = coerce_non_key_columns_to_numeric(cons_esp_df, export_key_cols)
cv_df             = coerce_non_key_columns_to_numeric(cv_df, export_key_cols)
cvnc_df           = coerce_non_key_columns_to_numeric(cvnc_df, export_key_cols)
comb_pond_df      = coerce_non_key_columns_to_numeric(comb_pond_df, export_key_cols)
ce_pond_df        = coerce_non_key_columns_to_numeric(ce_pond_df, export_key_cols)
cv_pond_df        = coerce_non_key_columns_to_numeric(cv_pond_df, export_key_cols)
cvnc_pond_df      = coerce_non_key_columns_to_numeric(cvnc_pond_df, export_key_cols)
reservas_tec_df   = coerce_non_key_columns_to_numeric(reservas_tec_df, export_key_cols)
comb_empresa_df   = coerce_non_key_columns_to_numeric(comb_empresa_df, export_key_cols)
gen_forzada_df    = coerce_non_key_columns_to_numeric(gen_forzada_df, export_key_cols)

# Define acá cuántas columnas en blanco quieres entre bloques
block_specs = [
    (drop_key_cols(subtipo_or_df, export_key_cols), 3),      # 1 blanco después
    (drop_key_cols(centrales_or_df, export_key_cols), 2),    # 1 blanco después
    (drop_key_cols(barras_or_df, export_key_cols), 2),       # 2 blancos después
    (drop_key_cols(subtipos_plan_df, export_key_cols), 1),
    (drop_key_cols(centrales_prg_df, export_key_cols), 2),
    (drop_key_cols(barras_plan_df, export_key_cols), 2),
    (drop_key_cols(precio_po_df, export_key_cols), 1),
    (drop_key_cols(cons_esp_df, export_key_cols), 1),
    (drop_key_cols(cv_df, export_key_cols), 1),
    (drop_key_cols(cvnc_df, export_key_cols), 2),
    (drop_key_cols(comb_pond_df, export_key_cols), 1),
    (drop_key_cols(ce_pond_df, export_key_cols), 1),
    (drop_key_cols(cv_pond_df, export_key_cols), 1),
    (drop_key_cols(cvnc_pond_df, export_key_cols), 1),
    (drop_key_cols(reservas_tec_df, export_key_cols), 2),
    (drop_key_cols(gen_forzada_df, export_key_cols), 2), 
    (drop_key_cols(comb_empresa_df, export_key_cols), 0),    # último sin blancos
]

time_unificada_df = assemble_blocks_flexible(block_specs)

keys_display_df = add_date_parts(master_time_keys_df.copy())

final_sheet_df = pd.concat(
    [
        keys_display_df,
        blank_cols(max(len(keys_display_df), len(time_unificada_df)), 1),
        time_unificada_df
    ],
    axis=1
)

# -----------------------
# 7) Escribir Excel
# -----------------------
print("Directorio actual:", os.getcwd())
print("Ruta absoluta Excel:", os.path.abspath(EXCEL_PATH))
print("Existe archivo:", os.path.exists(EXCEL_PATH))

OUTPUT_TEST = "unificada_2026-03-299.xlsx"

write_df_to_new_excel(
    final_sheet_df,
    path=OUTPUT_TEST,
    sheet_name=SHEET_NAME,
    start_row=START_ROW,
    start_col=START_COL,
)

print(f"OK -> creado {OUTPUT_TEST}")
print("Filas base maestra:", len(master_time_keys_df))
print("Filas bloque final:", len(final_sheet_df))
```
